```python
import json, requests, warnings, os, zipfile, time, io, base64, tempfile, shutil
from pathlib import Path
from itertools import product, permutations
from datetime import datetime

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

import ipywidgets as widgets
from IPython.display import display, clear_output, HTML

from pymatgen.core import Structure, Composition
from pymatgen.core.periodic_table import Element
from pymatgen.io.vasp import Poscar, Vasprun
from pymatgen.io.cif import CifWriter

from matgl import load_model
from matgl.ext.ase import Relaxer

# =====================================================================================
# SUPPRESS ALL WARNINGS
# =====================================================================================
warnings.filterwarnings("ignore")
import logging
logging.getLogger('matgl').setLevel(logging.ERROR)
logging.getLogger('ase').setLevel(logging.ERROR)

# =====================================================================================
# APP IDENTIFICATION (ADDED FOR ISOLATION)
# =====================================================================================
APP_ID = "DEFECT_FF"

# -------------------------------------------------------------------------------------
# Directories
# -------------------------------------------------------------------------------------
UPLOAD_DIR = Path.home() / "nanohub_uploads"
UPLOAD_DIR.mkdir(parents=True, exist_ok=True)

MODELS_DIR = Path.home() / "models"
MODELS_DIR.mkdir(parents=True, exist_ok=True)

GITHUB_BULK_BASE = "https://raw.githubusercontent.com/msehabibur/nanohub-simple-jupyter-tool/master/bulk_contcar"

# -------------------------------------------------------------------------------------
# Defect building utilities
# -------------------------------------------------------------------------------------
def read_structure(p: Path) -> Structure:
    return Structure.from_file(str(p))

def get_central_atom_idx(struct: Structure, symbol: str, excluded_indices=None) -> int:
    if excluded_indices is None:
        excluded_indices = []
    center = struct.lattice.get_cartesian_coords([0.5, 0.5, 0.5])
    min_dist, best_idx = float("inf"), -1
    for i, site in enumerate(struct):
        if i in excluded_indices:
            continue
        if site.specie.symbol != symbol:
            continue
        d = np.linalg.norm(site.coords - center)
        if d < min_dist:
            min_dist, best_idx = d, i
    if best_idx < 0:
        raise ValueError(f"No available atom '{symbol}' found near the cell center.")
    return best_idx

def get_min_pbc_distance(struct: Structure, cart_coords: np.ndarray) -> float:
    frac = struct.lattice.get_fractional_coords(cart_coords)
    mind = float("inf")
    for site in struct:
        df = frac - site.frac_coords
        df -= np.round(df)
        d = np.linalg.norm(struct.lattice.get_cartesian_coords(df))
        if d < mind:
            mind = d
    return float(mind)

def add_interstitial(struct: Structure, symbol: str):
    best_pos, best_d = None, -1.0
    for f in product(np.linspace(0.3, 0.7, 7), repeat=3):
        cart = struct.lattice.get_cartesian_coords(f)
        d = get_min_pbc_distance(struct, cart)
        if d > best_d:
            best_d, best_pos = d, cart
    if best_pos is None:
        raise ValueError(f"Could not place interstitial {symbol}.")
    struct.append(symbol, best_pos, coords_are_cartesian=True)

def create_defect_structure(bulk: Structure, defect_str: str) -> Structure:
    s = bulk.copy()
    used_indices = []
    species_in_struct = sorted({site.specie.symbol for site in s})

    for raw in defect_str.split("+"):
        part = raw.strip()
        if not part:
            continue

        # Vacancy
        if part.startswith("V_"):
            host = part[2:]
            if host not in species_in_struct:
                raise ValueError(f"Vacancy target '{host}' not in structure. Species: {species_in_struct}")
            idx = get_central_atom_idx(s, host, excluded_indices=used_indices)
            s.remove_sites([idx])
            used_indices.append(idx)
            continue

        # Interstitial
        if part.endswith("_i"):
            elem = part[:-2]
            if not elem:
                raise ValueError(f"Invalid interstitial token: '{part}'")
            add_interstitial(s, elem)
            continue

        # Substitution
        if "_" in part:
            A, B = part.split("_", 1)
            if B not in species_in_struct:
                raise ValueError(f"Substitution site '{B}' not in structure. Species: {species_in_struct}")
            idx = get_central_atom_idx(s, B, excluded_indices=used_indices)
            s.replace(idx, A)
            used_indices.append(idx)
            continue

        raise ValueError(f"Unknown defect token: '{part}'")

    return s

def get_defect_type(defect_str: str) -> str:
    if "+" in defect_str:
        return "defect complex"
    elif defect_str.startswith("V_"):
        return "vacancy"
    elif defect_str.endswith("_i"):
        return "interstitial"
    elif "_" in defect_str:
        return "substitution"
    return "unknown"

# -------------------------------------------------------------------------------------
# Model download + load from GitHub
# -------------------------------------------------------------------------------------
GITHUB_MODEL_BASE = {
    'PBE': 'https://raw.githubusercontent.com/msehabibur/nanohub-simple-jupyter-tool/master/data/PES_PBE',
    'HSE': 'https://raw.githubusercontent.com/msehabibur/nanohub-simple-jupyter-tool/master/data/PES_HSE'
}

CANDIDATE_FILES = ["model.json", "model.pt", "state.pt"]

def charge_model_names(model_type: str) -> dict:
    model_type = model_type.upper()
    if model_type == "PBE":
        return {0:"pbe_neutral", +1:"pbe_p1", +2:"pbe_p2", -1:"pbe_m1", -2:"pbe_m2"}
    if model_type == "HSE":
        return {0:"hse_neutral", +1:"hse_p1", +2:"hse_p2", -1:"hse_m1", -2:"hse_m2"}
    raise ValueError("model_type must be 'PBE' or 'HSE'")

def download_model_dir(model_type: str, model_name: str, log=print) -> Path:
    raw_base = GITHUB_MODEL_BASE[model_type.upper()]
    local_dir = MODELS_DIR / model_name
    local_dir.mkdir(parents=True, exist_ok=True)
    
    if (local_dir / "model.json").exists():
        log(f"‚úÖ {model_name} already cached")
        return local_dir
    
    log(f"‚¨áÔ∏è Downloading {model_name}...")
    
    for fn in CANDIDATE_FILES:
        url = f"{raw_base}/{model_name}/{fn}"
        dst = local_dir / fn
        
        if dst.exists():
            continue
        
        try:
            r = requests.get(url, timeout=120)
            if r.status_code == 200 and r.content:
                dst.write_bytes(r.content)
                size_kb = len(r.content) / 1024
                log(f"   ‚úì {fn} ({size_kb:.1f} KB)")
        except:
            pass
    
    if not (local_dir / "model.json").exists():
        raise FileNotFoundError(f"model.json not found for {model_name}")
    
    return local_dir

def load_relaxer(model_type: str, model_name: str, relax_cell: bool, log=print) -> Relaxer:
    try:
        local_dir = download_model_dir(model_type, model_name, log=log)
        log(f"Loading {model_name}...")
        model = load_model(str(local_dir))
        log(f"‚úÖ Loaded {model_name}")
        return Relaxer(model, relax_cell=relax_cell, optimizer="FIRE")
    except Exception as e:
        log(f"‚ùå Failed to load {model_name}: {e}")
        raise RuntimeError(f"Failed to load {model_name}: {e}")

def fetch_bulk_from_github(composition: str, log=print) -> Path:
    filename = f"{composition}.vasp"
    url = f"{GITHUB_BULK_BASE}/{filename}"
    dst = UPLOAD_DIR / filename
    
    log(f"Downloading {composition} from GitHub...")
    try:
        r = requests.get(url, timeout=30)
        if r.status_code == 200:
            dst.write_bytes(r.content)
            log(f"Successfully downloaded: {filename}")
            return dst
        else:
            raise Exception(f"HTTP {r.status_code}")
    except Exception as e:
        raise Exception(f"Failed to fetch: {e}")

# -------------------------------------------------------------------------------------
# Simple Structure Optimization (for uploaded files)
# -------------------------------------------------------------------------------------
def optimize_single_structure(
    structure_file: str,
    model_type: str,
    fmax: float,
    steps: int,
    relax_cell: bool,
    progress_cb=None,
    log=print,
):
    """Optimize a single uploaded structure"""
    
    charges = [-2, -1, 0, +1, +2]
    model_map = charge_model_names(model_type)
    
    total_steps = len(charges)
    done = 0
    
    def bump():
        nonlocal done
        done += 1
        if progress_cb:
            progress_cb(done, total_steps)
    
    log(f"Model: {model_type} | Charges: {charges} | steps={steps} | fmax={fmax}")
    
    # Read structure
    log(f"Reading structure from: {structure_file}")
    struct = read_structure(Path(structure_file))
    composition = struct.composition.reduced_formula
    
    log(f"Structure: {composition} ({len(struct)} atoms)")
    
    # IMPORTANT: Perturb structure to ensure optimization runs
    log("Perturbing atomic positions by 0.1 √Ö...")
    struct.perturb(0.10)
    log("Structure perturbed - ready for optimization")
    
    # Load relaxers
    log("Loading MatGL relaxers...")
    relaxers = {}
    for q in charges:
        name = model_map[q]
        try:
            rlx = load_relaxer(model_type, name, relax_cell=(relax_cell if q == 0 else False), log=log)
            relaxers[q] = rlx
            log(f"Loaded model for q={q:+d}: {name}")
        except Exception as e:
            log(f"Could not load model for q={q:+d}: {e}")
    
    if 0 not in relaxers:
        raise RuntimeError("Neutral model required but not loaded")
    
    # Optimize for all charge states
    energies = {}
    structures_dict = {}
    trajectories_dict = {}
    
    for q in charges:
        if q not in relaxers:
            log(f"Skipping q={q:+d} (model missing)")
            bump()
            continue
        
        relaxer = relaxers[q]
        relaxer.relax_cell = (relax_cell if q == 0 else False)
        
        try:
            log(f"Starting optimization for q={q:+d}...")
            log(f"  Max steps: {steps}, Fmax: {fmax} eV/√Ö")
            
            # Make a fresh copy for each charge state
            struct_copy = struct.copy()
            
            res = relaxer.relax(struct_copy, fmax=fmax, steps=steps)
            
            E_initial = float(res["trajectory"].energies[0])
            E_final = float(res["trajectory"].energies[-1])
            n_steps = len(res["trajectory"].energies)
            
            energies[q] = E_final
            
            # Store structure
            structures_dict[f"{composition}_q{q:+d}"] = str(CifWriter(res["final_structure"]))
            
            # Store trajectory
            traj = [float(e) for e in res["trajectory"].energies]
            trajectories_dict[f"q={q:+d}"] = traj
            
            log(f"‚úì q={q:+d}: {E_final:.6f} eV")
            log(f"  Initial energy: {E_initial:.6f} eV")
            log(f"  Energy change: {E_final - E_initial:.6f} eV")
            log(f"  Converged in {n_steps} steps")
            
        except Exception as e:
            log(f"‚úó Optimization failed q={q:+d}: {e}")
            import traceback
            traceback.print_exc()
        
        bump()
    
    return composition, energies, structures_dict, trajectories_dict

# -------------------------------------------------------------------------------------
# MLFF Optimization Engine (for defects)
# -------------------------------------------------------------------------------------
def run_mlff_optimization(
    bulk_file: str,
    defects_list: list,
    model_type: str,
    fmax: float,
    steps: int,
    relax_bulk_cell: bool,
    progress_cb=None,
    log=print,
):
    charges = [-2, -1, 0, +1, +2]
    model_map = charge_model_names(model_type)
    
    total_steps = len(defects_list) * len(charges) + 1
    done = 0
    
    def bump():
        nonlocal done
        done += 1
        if progress_cb:
            progress_cb(done, total_steps)
    
    log(f"Model: {model_type} | Charges: {charges} | steps={steps} | fmax={fmax}")
    
    bulk_initial = read_structure(Path(bulk_file))
    
    log("Loading MatGL relaxers...")
    relaxers = {}
    for q in charges:
        name = model_map[q]
        try:
            rlx = load_relaxer(model_type, name, relax_cell=(relax_bulk_cell if q == 0 else False), log=log)
            relaxers[q] = rlx
            log(f"Loaded model for q={q:+d}: {name}")
        except Exception as e:
            log(f"Could not load model for q={q:+d}: {e}")
    
    if 0 not in relaxers:
        raise RuntimeError("Neutral model required but not loaded")
    
    log("Relaxing bulk (q=0)...")
    bulk_res = relaxers[0].relax(bulk_initial.copy(), fmax=fmax, steps=steps)
    bulk_opt = bulk_res["final_structure"]
    E_bulk = float(bulk_res["trajectory"].energies[-1])
    log(f"Bulk energy: {E_bulk:.6f} eV")
    bump()
    
    energies_dict = {"bulk": E_bulk}
    structures_dict = {}
    trajectories_dict = {"Bulk": [float(e) for e in bulk_res["trajectory"].energies]}
    structures_dict["bulk"] = str(CifWriter(bulk_opt))
    
    log(f"Processing {len(defects_list)} defects...")
    for dstr in defects_list:
        energies_dict[dstr] = {}
        
        try:
            sdef = create_defect_structure(bulk_opt, dstr)
            sdef.perturb(0.10)
        except Exception as e:
            log(f"Build failed for {dstr}: {e}")
            for _ in charges:
                bump()
            continue
        
        for q in charges:
            if q not in relaxers:
                log(f"Skipping {dstr} q={q:+d} (model missing)")
                bump()
                continue
            
            relaxer = relaxers[q]
            relaxer.relax_cell = False
            try:
                res = relaxer.relax(sdef.copy(), fmax=fmax, steps=steps)
                E_final = float(res["trajectory"].energies[-1])
                energies_dict[dstr][q] = E_final
                structures_dict[f"{dstr}_q{q:+d}"] = str(CifWriter(res["final_structure"]))
                traj = [float(e) for e in res["trajectory"].energies]
                trajectories_dict[f"{dstr} (q={q:+d})"] = traj
                log(f"{dstr} (q={q:+d}): {E_final:.6f} eV")
            except Exception as e:
                log(f"Relax failed {dstr} (q={q:+d}): {e}")
            bump()
    
    return energies_dict, structures_dict, trajectories_dict, bulk_opt

def create_energy_table(energies_dict, composition, defects_list):
    charges = [-2, -1, 0, +1, +2]
    E_bulk = energies_dict["bulk"]
    
    rows = []
    for defect in defects_list:
        if defect not in energies_dict:
            continue
        
        row = {
            "Compound": composition,
            "Defect": defect,
            "Type": get_defect_type(defect),
            "E_bulk": E_bulk,
        }
        
        for q in charges:
            q_label = ['q=-2', 'q=-1', 'q=0', 'q=+1', 'q=+2'][charges.index(q)]
            row[f"E_{q_label}"] = energies_dict[defect].get(q, np.nan)
        
        rows.append(row)
    
    return pd.DataFrame(rows)

def plot_energy_trajectories(trajectories_dict):
    plt.style.use('default')
    fig, ax = plt.subplots(figsize=(12, 7), facecolor='white')
    
    colors = plt.cm.tab20(np.linspace(0, 1, len(trajectories_dict)))
    
    for (label, energies), color in zip(trajectories_dict.items(), colors):
        steps = range(len(energies))
        ax.plot(steps, energies, label=label, linewidth=2.5, alpha=0.85, color=color, 
               marker='o', markersize=3, markevery=max(1, len(energies)//20))
    
    ax.set_xlabel('Optimization Step', fontsize=22, fontweight='bold', labelpad=10)
    ax.set_ylabel('Energy (eV)', fontsize=22, fontweight='bold', labelpad=10)
    ax.set_title('MLFF Optimization: Energy Trajectories', 
                 fontsize=18, fontweight='bold', pad=15)
    ax.legend(loc='best', fontsize=14, framealpha=0.95, shadow=True, ncol=2)
    ax.grid(True, alpha=0.3, linestyle='--', linewidth=1)
    ax.tick_params(labelsize=22, width=1.5, length=6)
    
    plt.tight_layout()
    return fig

def create_download_link(content, filename, content_type='text/plain'):
    if isinstance(content, str):
        content = content.encode()
    
    b64 = base64.b64encode(content).decode()
    href = f'data:{content_type};base64,{b64}'
    
    html = f'''
    <a href="{href}" 
       download="{filename}" 
       style="display:none" 
       id="dl_{filename.replace(".", "_").replace(" ", "_")}"></a>
    <script>
        document.getElementById("dl_{filename.replace(".", "_").replace(" ", "_")}").click();
    </script>
    '''
    return HTML(html)

# =====================================================================================
# GUI IMPLEMENTATION WITH UNIQUE VARIABLE NAMES
# =====================================================================================

def _decode_bytes(b: bytes) -> str:
    try:
        return b.decode("utf-8")
    except UnicodeDecodeError:
        return b.decode("latin-1", errors="replace")

defect_ff_header = widgets.HTML("""
<style>
    .defect-header {
        background: linear-gradient(135deg, #9ca3af 0%, #6b7280 100%);
        padding: 20px;
        border-radius: 15px;
        box-shadow: 0 10px 30px rgba(0,0,0,0.2);
        margin-bottom: 15px;
    }
    .defect-title {
        color: white;
        font-size: 24px;
        font-weight: 800;
        margin: 0;
        text-shadow: 2px 2px 4px rgba(0,0,0,0.3);
        text-align: center;
    }
    .defect-subtitle {
        color: #f3f4f6;
        font-size: 13px;
        margin-top: 8px;
        text-align: center;
    }
    .widget-button:hover {
        opacity: 0.85;
        transform: translateY(-2px);
        transition: all 0.2s ease;
    }
    .widget-button:active {
        transform: translateY(0px);
        opacity: 0.95;
    }
</style>
<div class="defect-header">
    <div class="defect-title">üß™ DeFecT-FF Calculator</div>
    <div class="defect-subtitle">
        Machine Learning Force Field Optimization for Structures
    </div>
</div>
""")

defect_ff_compositions = [
    "Cd0.25Zn0.75Te", "Cd0.50Zn0.50Te", "Cd0.75Zn0.25Te",
    "CdSe", "CdSe0.06Te0.94", "CdSe0.12Te0.88",
    "CdSe0.20Te0.80", "CdSe0.25Te0.75", "CdSe0.50Te0.50",
    "CdSe0.75Te0.25", "CdTe", "ZnTe"
]

defect_ff_workflow_label = widgets.HTML("<h4 style='color:#4b5563;font-weight:700;margin:5px 0;'>Step 1: Choose Workflow</h4>")

defect_ff_workflow_upload_btn = widgets.Button(
    description="üì§ Upload Structure",
    button_style="",
    layout=widgets.Layout(width='180px', height='45px'),
    style={'font_weight': 'bold'}
)

defect_ff_workflow_select_btn = widgets.Button(
    description="üîç Build Defects",
    button_style="",
    layout=widgets.Layout(width='180px', height='45px'),
    style={'font_weight': 'bold'}
)

defect_ff_upload_structure = widgets.FileUpload(
    accept=".vasp,.POSCAR,.cif",
    multiple=False,
    description="Structure:",
    style={'description_width': '80px'},
    layout=widgets.Layout(width='100%')
)

defect_ff_composition_dropdown = widgets.Dropdown(
    options=["Select composition..."] + defect_ff_compositions,
    value="Select composition...",
    description="Composition:",
    style={'description_width': '90px'},
    layout=widgets.Layout(width='100%')
)

defect_ff_syntax_card = widgets.HTML("""
<div style="padding:10px;border-radius:8px;background:#f5f7fa;
            border-left:4px solid #9ca3af;margin-bottom:10px;font-size:11px;">
  <b>Defect Syntax:</b> <code>V_Cd</code> (vacancy) | <code>As_Te</code> (substitution) | 
  <code>Cu_i</code> (interstitial) | <code>Cl_Te+V_Cd</code> (complex)
</div>
""")

defect_ff_defect_text = widgets.Textarea(
    value="V_Cd\nAs_Te\nCu_i\nCl_Te+V_Cd",
    placeholder="One defect per line...",
    layout=widgets.Layout(width="100%", height="140px")
)

defect_ff_model_toggle = widgets.ToggleButtons(
    options=["PBE", "HSE"],
    value="PBE",
    description="Model:",
    style={"description_width": "60px"},
    button_style='',
    layout=widgets.Layout(width='100%')
)

defect_ff_steps_slider = widgets.IntSlider(
    value=200, min=1, max=2000, step=10,
    description="Steps:",
    style={"description_width": "60px"},
    layout=widgets.Layout(width='100%')
)

defect_ff_fmax_slider = widgets.FloatSlider(
    value=0.05, min=0.001, max=0.2, step=0.005,
    description="Fmax:",
    readout_format=".3f",
    style={"description_width": "60px"},
    layout=widgets.Layout(width='100%')
)

defect_ff_relax_cell = widgets.Checkbox(
    value=True,
    description="Relax cell",
    style={'description_width': '100px'}
)

defect_ff_run_button = widgets.Button(
    description="üöÄ Run Optimization",
    button_style="",
    layout=widgets.Layout(width='100%', height='50px'),
    style={'font_weight': 'bold'}
)

defect_ff_download_btn = widgets.Button(
    description="üì• Download Results",
    button_style='',
    layout=widgets.Layout(width='100%', height='50px'),
    style={'font_weight': 'bold'},
    disabled=True
)

defect_ff_progress = widgets.IntProgress(
    value=0, min=0, max=1,
    description="Progress:",
    style={"description_width": "60px", 'bar_color': '#9ca3af'},
    layout=widgets.Layout(width='100%')
)

defect_ff_status = widgets.HTML("<div style='padding:8px;background:#f9fafb;border-left:4px solid #9ca3af;border-radius:6px;font-size:12px;'>"
                      "<b>Status:</b> Ready</div>")

defect_ff_log_out = widgets.Output(layout={"border": "2px solid #e2e8f0", "border_radius": "8px",
                                  "padding": "10px", "height": "300px", "overflow": "auto",
                                  "background": "#1a202c"})
defect_ff_log_out._dom_classes = [f'{APP_ID}_log']

defect_ff_plot_out = widgets.Output(layout={"border": "2px solid #e2e8f0", "border_radius": "8px",
                                   "padding": "10px", "background": "white"})
defect_ff_plot_out._dom_classes = [f'{APP_ID}_plot']

defect_ff_workflow_mode = None
defect_ff_global_results = {}

def defect_ff_parse_defects(text: str) -> list:
    lines = []
    for ln in text.splitlines():
        ln = ln.strip()
        if not ln or ln.startswith("#"):
            continue
        lines.append(ln)
    return lines

def defect_ff_set_progress(cur, total):
    defect_ff_progress.max = max(1, int(total))
    defect_ff_progress.value = min(int(cur), defect_ff_progress.max)

def defect_ff_on_workflow_upload_clicked(_):
    global defect_ff_workflow_mode
    defect_ff_workflow_mode = 'upload'
    defect_ff_build_main_ui()

def defect_ff_on_workflow_select_clicked(_):
    global defect_ff_workflow_mode
    defect_ff_workflow_mode = 'select'
    defect_ff_build_main_ui()

defect_ff_workflow_upload_btn.on_click(defect_ff_on_workflow_upload_clicked)
defect_ff_workflow_select_btn.on_click(defect_ff_on_workflow_select_clicked)

def defect_ff_on_run_clicked(_):
    global defect_ff_global_results
    
    defect_ff_log_out.clear_output(wait=True)
    with defect_ff_log_out:
        
        if defect_ff_workflow_mode == 'upload':
            print("\033[96m[DEBUG] Entering UPLOAD mode\033[0m")
            
            # UPLOAD MODE: Just optimize the uploaded structure
            if not defect_ff_upload_structure.value:
                print("\033[91m[ERROR] Please upload a structure file\033[0m")
                return
            
            print(f"\033[96m[DEBUG] File uploaded: {list(defect_ff_upload_structure.value.keys())}\033[0m")
            
            struct_name = list(defect_ff_upload_structure.value.keys())[0]
            struct_path = UPLOAD_DIR / "uploaded_structure_ff.vasp"
            struct_path.write_text(_decode_bytes(defect_ff_upload_structure.value[struct_name]["content"]))
            
            print(f"\033[92m[INFO] Uploaded: {struct_name}\033[0m")
            print(f"\033[92m[INFO] Saved to: {struct_path}\033[0m")
            print(f"\033[94m[INFO] Model: {defect_ff_model_toggle.value}\033[0m")
            print("=" * 80)
            
            defect_ff_set_progress(0, 5)
            defect_ff_status.value = ("<div style='padding:8px;background:#f3f4f6;border-left:4px solid #9ca3af;"
                           "border-radius:6px;font-size:12px;'><b>Status:</b> Optimizing...</div>")
            
            print("\033[96m\n=== STRUCTURE OPTIMIZATION ===\033[0m")
            try:
                composition, energies, structures_dict, trajectories_dict = optimize_single_structure(
                    structure_file=str(struct_path),
                    model_type=defect_ff_model_toggle.value,
                    fmax=float(defect_ff_fmax_slider.value),
                    steps=int(defect_ff_steps_slider.value),
                    relax_cell=bool(defect_ff_relax_cell.value),
                    progress_cb=defect_ff_set_progress,
                    log=lambda msg: print(f"\033[93m[OPT] {msg}\033[0m")
                )
                print("\033[92m[SUCCESS] Optimization complete\033[0m")
                
            except Exception as e:
                print(f"\033[91m[ERROR] Optimization failed: {e}\033[0m")
                import traceback
                traceback.print_exc()
                defect_ff_status.value = ("<div style='padding:8px;background:#fee2e2;border-left:4px solid #9ca3af;"
                               "border-radius:6px;font-size:12px;'><b>Status:</b> Failed</div>")
                return
            
            # Energy summary
            print("\033[96m\n=== ENERGY SUMMARY ===\033[0m")
            print(f"Composition: {composition}")
            for q, E in sorted(energies.items()):
                print(f"  q={q:+d}: {E:.6f} eV")
            
            # Plotting
            print("\033[96m\n=== GENERATING PLOT ===\033[0m")
    
    # PLOT OUTSIDE log_out context to ensure it displays in plot_out
    if defect_ff_workflow_mode == 'upload':
        defect_ff_plot_out.clear_output(wait=True)
        with defect_ff_plot_out:
            try:
                fig = plot_energy_trajectories(trajectories_dict)
                plt.show()
            except Exception as e:
                print(f"Plot error: {e}")
        
        with defect_ff_log_out:
            print("\033[92m[SUCCESS] Plot generated\033[0m")
            
            # Store results
            defect_ff_global_results = {
                'composition': composition,
                'energies': energies,
                'structures_dict': structures_dict,
                'trajectories_dict': trajectories_dict,
                'figure': fig,
                'mode': 'upload'
            }
            
    elif defect_ff_workflow_mode == 'select':
        with defect_ff_log_out:
            print("\033[96m[DEBUG] Entering DEFECT mode\033[0m")
            
            # DEFECT MODE: Build defects and optimize
            if defect_ff_composition_dropdown.value == "Select composition...":
                print("\033[91m[ERROR] Please select composition\033[0m")
                return
            
            composition = defect_ff_composition_dropdown.value
            try:
                bulk_path = fetch_bulk_from_github(
                    composition,
                    log=lambda msg: print(f"\033[93m[FETCH] {msg}\033[0m")
                )
            except Exception as e:
                print(f"\033[91m[ERROR] {e}\033[0m")
                return
            
            defects_list = defect_ff_parse_defects(defect_ff_defect_text.value)
            if not defects_list:
                print("\033[91m[ERROR] Please enter defects\033[0m")
                return
            
            total_steps = len(defects_list) * 5 + 1
            defect_ff_set_progress(0, total_steps)
            defect_ff_status.value = ("<div style='padding:8px;background:#f3f4f6;border-left:4px solid #9ca3af;"
                           "border-radius:6px;font-size:12px;'><b>Status:</b> Running...</div>")
            
            print(f"\033[92m[INFO] Composition: {composition}\033[0m")
            print(f"\033[92m[INFO] Defects: {defects_list}\033[0m")
            print(f"\033[94m[INFO] Model: {defect_ff_model_toggle.value}\033[0m")
            print("=" * 80)
            
            print("\033[96m\n=== MLFF OPTIMIZATION ===\033[0m")
            try:
                energies_dict, structures_dict, trajectories_dict, bulk_opt = run_mlff_optimization(
                    bulk_file=str(bulk_path),
                    defects_list=defects_list,
                    model_type=defect_ff_model_toggle.value,
                    fmax=float(defect_ff_fmax_slider.value),
                    steps=int(defect_ff_steps_slider.value),
                    relax_bulk_cell=bool(defect_ff_relax_cell.value),
                    progress_cb=defect_ff_set_progress,
                    log=lambda msg: print(f"\033[93m[MLFF] {msg}\033[0m")
                )
                print("\033[92m[SUCCESS] MLFF optimization complete\033[0m")
                
            except Exception as e:
                print(f"\033[91m[ERROR] MLFF failed: {e}\033[0m")
                import traceback
                traceback.print_exc()
                return
            
            print("\033[96m\n=== CREATING ENERGY TABLE ===\033[0m")
            energy_df = create_energy_table(energies_dict, composition, defects_list)
            print("\033[92m[SUCCESS] Energy table created\033[0m")
            
            print("\nEnergy Table (preview):")
            preview_cols = ['Compound', 'Defect', 'Type', 'E_bulk', 'E_q=-2', 'E_q=-1', 'E_q=0', 'E_q=+1', 'E_q=+2']
            if all(col in energy_df.columns for col in preview_cols):
                preview_df = energy_df[preview_cols].copy()
                energy_cols = ['E_bulk', 'E_q=-2', 'E_q=-1', 'E_q=0', 'E_q=+1', 'E_q=+2']
                for col in energy_cols:
                    if col in preview_df.columns:
                        preview_df[col] = preview_df[col].apply(lambda x: f"{x:.2f}" if pd.notna(x) else "")
                
                pd.set_option('display.max_columns', None)
                pd.set_option('display.width', 120)
                pd.set_option('display.colheader_justify', 'left')
                print(preview_df.to_string(index=False))
                pd.reset_option('display.max_columns')
                pd.reset_option('display.width')
                pd.reset_option('display.colheader_justify')
            
            print("\033[96m\n=== GENERATING PLOTS ===\033[0m")
        
        # PLOT OUTSIDE log_out context
        defect_ff_plot_out.clear_output(wait=True)
        with defect_ff_plot_out:
            try:
                fig = plot_energy_trajectories(trajectories_dict)
                plt.show()
            except Exception as e:
                print(f"Plot error: {e}")
        
        with defect_ff_log_out:
            print("\033[92m[SUCCESS] Plots generated\033[0m")
            
            defect_ff_global_results = {
                'composition': composition,
                'energies_dict': energies_dict,
                'structures_dict': structures_dict,
                'trajectories_dict': trajectories_dict,
                'energy_df': energy_df,
                'figure': fig,
                'defects_list': defects_list,
                'model_type': defect_ff_model_toggle.value,
                'mode': 'defect'
            }
    
    else:
        with defect_ff_log_out:
            print(f"\033[91m[ERROR] Unknown workflow mode: {defect_ff_workflow_mode}\033[0m")
            return
    
    # Enable download
    defect_ff_download_btn.disabled = False
    
    with defect_ff_log_out:
        defect_ff_status.value = ("<div style='padding:8px;background:#f9fafb;border-left:4px solid #9ca3af;"
                       "border-radius:6px;font-size:12px;'><b>Status:</b> Complete! ‚úì</div>")
        print("\n" + "=" * 80)
        print("\033[92m[COMPLETE] All calculations finished!\033[0m")

defect_ff_run_button.on_click(defect_ff_on_run_clicked)

def defect_ff_on_download_clicked(b):
    if not defect_ff_global_results:
        return
    
    zip_buffer = io.BytesIO()
    
    if defect_ff_global_results['mode'] == 'upload':
        # Simple upload mode
        with zipfile.ZipFile(zip_buffer, 'w', zipfile.ZIP_DEFLATED) as zf:
            # Energies CSV
            energy_data = []
            for q, E in sorted(defect_ff_global_results['energies'].items()):
                energy_data.append({'Charge': f"q={q:+d}", 'Energy (eV)': E})
            df = pd.DataFrame(energy_data)
            zf.writestr("energies.csv", df.to_csv(index=False))
            
            # Structures
            for name, cif in defect_ff_global_results['structures_dict'].items():
                zf.writestr(f"structures/{name}.cif", cif)
            
            # Plot
            img_buffer = io.BytesIO()
            defect_ff_global_results['figure'].savefig(img_buffer, format='png', dpi=300, bbox_inches='tight', facecolor='white')
            img_buffer.seek(0)
            zf.writestr("energy_trajectory.png", img_buffer.read())
            
            # README
            readme = f"""
# Structure Optimization Results (MLFF)
Composition: {defect_ff_global_results['composition']}
Model: {defect_ff_model_toggle.value}
Date: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}

## Files:
- energies.csv: Energies for all charge states
- structures/: Optimized geometries (CIF format)
- energy_trajectory.png: Energy vs optimization step plot
"""
            zf.writestr("README.txt", readme)
    
    else:
        # Defect mode
        with zipfile.ZipFile(zip_buffer, 'w', zipfile.ZIP_DEFLATED) as zf:
            zf.writestr("energy_table.csv", defect_ff_global_results['energy_df'].to_csv(index=False))
            
            for name, cif in defect_ff_global_results['structures_dict'].items():
                zf.writestr(f"structures/{name}.cif", cif)
            
            img_buffer = io.BytesIO()
            defect_ff_global_results['figure'].savefig(img_buffer, format='png', dpi=300, bbox_inches='tight', facecolor='white')
            img_buffer.seek(0)
            zf.writestr("energy_trajectories.png", img_buffer.read())
            
            readme = f"""
# Defect Energy Calculation Results (MLFF)
Composition: {defect_ff_global_results['composition']}
Defects: {', '.join(defect_ff_global_results['defects_list'])}
Model: {defect_ff_global_results['model_type']}
Date: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}

## Files:
- energy_table.csv: Defect energies for all charge states
- structures/: Optimized geometries (CIF format)
- energy_trajectories.png: Energy trajectory plot
"""
            zf.writestr("README.txt", readme)
    
    zip_buffer.seek(0)
    display(create_download_link(
        zip_buffer.read(),
        f"{defect_ff_global_results['composition']}_MLFF_RESULTS.zip",
        "application/zip"
    ))

defect_ff_download_btn.on_click(defect_ff_on_download_clicked)

def defect_ff_build_main_ui():
    clear_output()
    
    display(defect_ff_header)
    
    if defect_ff_workflow_mode == 'upload':
        # UPLOAD MODE: Simple interface
        left_col = widgets.VBox([
            defect_ff_workflow_label,
            widgets.HBox([defect_ff_workflow_upload_btn, defect_ff_workflow_select_btn]),
            widgets.HTML("<h4 style='color:#4b5563;font-weight:700;margin:10px 0 5px 0;'>Upload Structure</h4>"),
            defect_ff_upload_structure,
            widgets.HTML("<h4 style='color:#4b5563;font-weight:700;margin:10px 0 5px 0;'>Settings</h4>"),
            defect_ff_model_toggle,
            defect_ff_steps_slider,
            defect_ff_fmax_slider,
            defect_ff_relax_cell,
            widgets.HTML("<h4 style='color:#4b5563;font-weight:700;margin:10px 0 5px 0;'>Run</h4>"),
            defect_ff_run_button,
            defect_ff_progress,
            defect_ff_status,
            widgets.HTML("<h4 style='color:#4b5563;font-weight:700;margin:10px 0 5px 0;'>Download</h4>"),
            defect_ff_download_btn,
        ], layout=widgets.Layout(width='48%', padding='5px'))
        
    else:
        # DEFECT MODE: Full interface
        left_col = widgets.VBox([
            defect_ff_workflow_label,
            widgets.HBox([defect_ff_workflow_upload_btn, defect_ff_workflow_select_btn]),
            defect_ff_composition_dropdown,
            widgets.HTML("<h4 style='color:#4b5563;font-weight:700;margin:10px 0 5px 0;'>Defects</h4>"),
            defect_ff_defect_text,
            defect_ff_syntax_card,
            widgets.HTML("<h4 style='color:#4b5563;font-weight:700;margin:10px 0 5px 0;'>Settings</h4>"),
            defect_ff_model_toggle,
            defect_ff_steps_slider,
            defect_ff_fmax_slider,
            defect_ff_relax_cell,
            widgets.HTML("<h4 style='color:#4b5563;font-weight:700;margin:10px 0 5px 0;'>Run</h4>"),
            defect_ff_run_button,
            defect_ff_progress,
            defect_ff_status,
            widgets.HTML("<h4 style='color:#4b5563;font-weight:700;margin:10px 0 5px 0;'>Download</h4>"),
            defect_ff_download_btn,
        ], layout=widgets.Layout(width='48%', padding='5px'))
    
    right_col = widgets.VBox([
        widgets.HTML("<h4 style='color:#4b5563;font-weight:700;margin:0 0 5px 0;'>üìü Execution Log</h4>"),
        defect_ff_log_out,
        widgets.HTML("<h4 style='color:#4b5563;font-weight:700;margin:10px 0 5px 0;'>üìä Results</h4>"),
        defect_ff_plot_out,
    ], layout=widgets.Layout(width='48%', padding='5px'))
    
    display(widgets.HBox([left_col, right_col], layout=widgets.Layout(justify_content='space-between')))

display(defect_ff_header)
display(defect_ff_workflow_label)
display(widgets.HBox([defect_ff_workflow_upload_btn, defect_ff_workflow_select_btn],
                    layout=widgets.Layout(justify_content='center')))
```

```python
import json, requests, warnings, os, zipfile, time, io, base64, tempfile, shutil, sys
from pathlib import Path
from itertools import product
from datetime import datetime
import random

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

import ipywidgets as widgets
from IPython.display import display, clear_output, HTML

from pymatgen.core import Structure, Composition
from pymatgen.core.periodic_table import Element
from pymatgen.io.vasp import Poscar
from pymatgen.io.cif import CifWriter

from matgl import load_model
from matgl.ext.ase import Relaxer

# =====================================================================================
# SUPPRESS WARNINGS
# =====================================================================================
warnings.filterwarnings("ignore")
import logging
logging.getLogger('matgl').setLevel(logging.ERROR)
logging.getLogger('ase').setLevel(logging.ERROR)

# =====================================================================================
# APP IDENTIFICATION (ADDED FOR ISOLATION)
# =====================================================================================
DEFECT_DB_APP_ID = "DEFECT_DB"

# =====================================================================================
# APP MODE DETECTION
# =====================================================================================
def defect_db_is_app_mode():
    """Check if running in nanoHUB app mode"""
    return 'app' in sys.argv or '--appmode' in sys.argv

defect_db_APP_MODE = defect_db_is_app_mode()

# =====================================================================================
# CONFIGURATION
# =====================================================================================
DEFECT_DB_UPLOAD_DIR = Path.home() / "nanohub_uploads"
DEFECT_DB_UPLOAD_DIR.mkdir(parents=True, exist_ok=True)

DEFECT_DB_MODELS_DIR = Path.home() / "models"
DEFECT_DB_MODELS_DIR.mkdir(parents=True, exist_ok=True)

DEFECT_DB_GITHUB_BULK_BASE = "https://raw.githubusercontent.com/msehabibur/nanohub-simple-jupyter-tool/master/bulk_contcar"

DEFECT_DB_GITHUB_MODEL_BASE = {
    'PBE': 'https://raw.githubusercontent.com/msehabibur/nanohub-simple-jupyter-tool/master/data/PES_PBE',
    'HSE': 'https://raw.githubusercontent.com/msehabibur/nanohub-simple-jupyter-tool/master/data/PES_HSE'
}

DEFECT_DB_CANDIDATE_FILES = ["model.json", "model.pt", "state.pt"]

# =====================================================================================
# DEFECT BUILDING UTILITIES
# =====================================================================================

def defect_db_read_structure(p: Path) -> Structure:
    return Structure.from_file(str(p))

def defect_db_get_central_atom_idx(struct: Structure, symbol: str, excluded_indices=None) -> int:
    if excluded_indices is None:
        excluded_indices = []
    center = struct.lattice.get_cartesian_coords([0.5, 0.5, 0.5])
    min_dist, best_idx = float("inf"), -1
    for i, site in enumerate(struct):
        if i in excluded_indices:
            continue
        if site.specie.symbol != symbol:
            continue
        d = np.linalg.norm(site.coords - center)
        if d < min_dist:
            min_dist, best_idx = d, i
    if best_idx < 0:
        raise ValueError(f"No available atom '{symbol}' found near the cell center.")
    return best_idx

def defect_db_get_min_pbc_distance(struct: Structure, cart_coords: np.ndarray) -> float:
    frac = struct.lattice.get_fractional_coords(cart_coords)
    mind = float("inf")
    for site in struct:
        df = frac - site.frac_coords
        df -= np.round(df)
        d = np.linalg.norm(struct.lattice.get_cartesian_coords(df))
        if d < mind:
            mind = d
    return float(mind)

def defect_db_add_interstitial(struct: Structure, symbol: str):
    best_pos, best_d = None, -1.0
    for f in product(np.linspace(0.3, 0.7, 7), repeat=3):
        cart = struct.lattice.get_cartesian_coords(f)
        d = defect_db_get_min_pbc_distance(struct, cart)
        if d > best_d:
            best_d, best_pos = d, cart
    if best_pos is None:
        raise ValueError(f"Could not place interstitial {symbol}.")
    struct.append(symbol, best_pos, coords_are_cartesian=True)

def defect_db_create_defect_structure(bulk: Structure, defect_str: str) -> Structure:
    """Create defect structure from string like 'V_Cd', 'Cu_i', 'As_Te', 'Cl_Te+V_Cd'"""
    s = bulk.copy()
    used_indices = []
    species_in_struct = sorted({site.specie.symbol for site in s})

    for raw in defect_str.split("+"):
        part = raw.strip()
        if not part:
            continue

        # Vacancy
        if part.startswith("V_"):
            host = part[2:]
            if host not in species_in_struct:
                raise ValueError(f"Vacancy target '{host}' not in structure. Species: {species_in_struct}")
            idx = defect_db_get_central_atom_idx(s, host, excluded_indices=used_indices)
            s.remove_sites([idx])
            used_indices.append(idx)
            continue

        # Interstitial
        if part.endswith("_i"):
            elem = part[:-2]
            if not elem:
                raise ValueError(f"Invalid interstitial token: '{part}'")
            defect_db_add_interstitial(s, elem)
            continue

        # Substitution
        if "_" in part:
            A, B = part.split("_", 1)
            if B not in species_in_struct:
                raise ValueError(f"Substitution site '{B}' not in structure. Species: {species_in_struct}")
            idx = defect_db_get_central_atom_idx(s, B, excluded_indices=used_indices)
            s.replace(idx, A)
            used_indices.append(idx)
            continue

        raise ValueError(f"Unknown defect token: '{part}'")

    return s

def defect_db_get_defect_type(defect_str: str) -> str:
    if "+" in defect_str:
        return "defect complex"
    elif defect_str.startswith("V_"):
        return "vacancy"
    elif defect_str.endswith("_i"):
        return "interstitial"
    elif "_" in defect_str:
        return "substitution"
    return "unknown"

# =====================================================================================
# MODEL UTILITIES
# =====================================================================================

def defect_db_charge_model_names(model_type: str) -> dict:
    model_type = model_type.upper()
    if model_type == "PBE":
        return {0:"pbe_neutral", +1:"pbe_p1", +2:"pbe_p2", -1:"pbe_m1", -2:"pbe_m2"}
    if model_type == "HSE":
        return {0:"hse_neutral", +1:"hse_p1", +2:"hse_p2", -1:"hse_m1", -2:"hse_m2"}
    raise ValueError("model_type must be 'PBE' or 'HSE'")

def defect_db_download_model_dir(model_type: str, model_name: str, log=print) -> Path:
    raw_base = DEFECT_DB_GITHUB_MODEL_BASE[model_type.upper()]
    local_dir = DEFECT_DB_MODELS_DIR / model_name
    local_dir.mkdir(parents=True, exist_ok=True)
    
    if (local_dir / "model.json").exists():
        log(f"‚úÖ {model_name} already cached")
        return local_dir
    
    log(f"‚¨áÔ∏è Downloading {model_name}...")
    
    for fn in DEFECT_DB_CANDIDATE_FILES:
        url = f"{raw_base}/{model_name}/{fn}"
        dst = local_dir / fn
        
        if dst.exists():
            continue
        
        try:
            r = requests.get(url, timeout=120)
            if r.status_code == 200 and r.content:
                dst.write_bytes(r.content)
                size_kb = len(r.content) / 1024
                log(f"   ‚úì {fn} ({size_kb:.1f} KB)")
        except:
            pass
    
    if not (local_dir / "model.json").exists():
        raise FileNotFoundError(f"model.json not found for {model_name}")
    
    return local_dir

def defect_db_load_relaxer(model_type: str, model_name: str, relax_cell: bool, log=print) -> Relaxer:
    try:
        local_dir = defect_db_download_model_dir(model_type, model_name, log=log)
        log(f"Loading {model_name}...")
        model = load_model(str(local_dir))
        log(f"‚úÖ Loaded {model_name}")
        return Relaxer(model, relax_cell=relax_cell, optimizer="FIRE")
    except Exception as e:
        log(f"‚ùå Failed to load {model_name}: {e}")
        raise RuntimeError(f"Failed to load {model_name}: {e}")

def defect_db_fetch_bulk_from_github(composition: str, log=print) -> Path:
    filename = f"{composition}.vasp"
    url = f"{DEFECT_DB_GITHUB_BULK_BASE}/{filename}"
    dst = DEFECT_DB_UPLOAD_DIR / filename
    
    log(f"Downloading {composition} from GitHub...")
    try:
        r = requests.get(url, timeout=30)
        if r.status_code == 200:
            dst.write_bytes(r.content)
            log(f"Successfully downloaded: {filename}")
            return dst
        else:
            raise Exception(f"HTTP {r.status_code}")
    except Exception as e:
        raise Exception(f"Failed to fetch: {e}")

# =====================================================================================
# CONFIGURATION SAMPLING ENGINE
# =====================================================================================

def defect_db_parse_defect_string(defect_str: str) -> dict:
    """Parse defect string to extract type and elements"""
    defect_str = defect_str.strip()
    
    # Complex defect
    if "+" in defect_str:
        return {
            'type': 'complex',
            'defect_string': defect_str,
            'parts': [defect_db_parse_defect_string(p) for p in defect_str.split("+")]
        }
    
    # Vacancy: V_Cd
    if defect_str.startswith("V_"):
        return {
            'type': 'vacancy',
            'element': defect_str[2:],
            'defect_string': defect_str
        }
    
    # Interstitial: Cu_i
    if defect_str.endswith("_i"):
        return {
            'type': 'interstitial',
            'element': defect_str[:-2],
            'defect_string': defect_str
        }
    
    # Substitution: As_Te
    if "_" in defect_str:
        parts = defect_str.split("_", 1)
        return {
            'type': 'substitution',
            'element_add': parts[0],
            'element_replace': parts[1],
            'defect_string': defect_str
        }
    
    raise ValueError(f"Invalid defect string: {defect_str}")

def defect_db_generate_defect_configurations(bulk: Structure, defect_str: str, n_configs: int, 
                                   strategy: str = "random", log=print) -> list:
    """
    Generate multiple configurations of the same defect at different positions
    
    Args:
        bulk: Bulk structure
        defect_str: Defect string (e.g., 'V_Cd', 'Cu_i', 'As_Te', 'Cl_Te+V_Cd')
        n_configs: Number of configurations to generate
        strategy: 'random', 'systematic', or 'all'
    
    Returns:
        List of (Structure, label) tuples
    """
    
    defect_info = defect_db_parse_defect_string(defect_str)
    configs = []
    
    log(f"Generating configurations for: {defect_str}")
    log(f"Defect type: {defect_info['type']}")
    
    if defect_info['type'] == 'vacancy':
        # Generate vacancy at different atomic positions
        element = defect_info['element']
        indices = [i for i, site in enumerate(bulk) if site.specie.symbol == element]
        
        if not indices:
            raise ValueError(f"No {element} atoms found in structure")
        
        log(f"Found {len(indices)} possible {element} sites")
        
        # Select which sites to sample
        if strategy == "all" or n_configs >= len(indices):
            selected_indices = indices
        elif strategy == "random":
            selected_indices = random.sample(indices, min(n_configs, len(indices)))
        elif strategy == "systematic":
            step = max(1, len(indices) // n_configs)
            selected_indices = indices[::step][:n_configs]
        
        # Create configurations
        for idx in selected_indices:
            struct = bulk.copy()
            struct.remove_sites([idx])
            struct.perturb(0.05)  # Small perturbation
            configs.append((struct, f"{defect_str}_site{idx}"))
    
    elif defect_info['type'] == 'substitution':
        # Generate substitution at different positions
        element_replace = defect_info['element_replace']
        element_add = defect_info['element_add']
        
        indices = [i for i, site in enumerate(bulk) if site.specie.symbol == element_replace]
        
        if not indices:
            raise ValueError(f"No {element_replace} atoms found in structure")
        
        log(f"Found {len(indices)} possible {element_replace} sites")
        
        # Select sites
        if strategy == "all" or n_configs >= len(indices):
            selected_indices = indices
        elif strategy == "random":
            selected_indices = random.sample(indices, min(n_configs, len(indices)))
        elif strategy == "systematic":
            step = max(1, len(indices) // n_configs)
            selected_indices = indices[::step][:n_configs]
        
        # Create configurations
        for idx in selected_indices:
            struct = bulk.copy()
            struct.replace(idx, element_add)
            struct.perturb(0.05)
            configs.append((struct, f"{defect_str}_site{idx}"))
    
    elif defect_info['type'] == 'interstitial':
        # Generate interstitial at different positions in the cell
        element = defect_info['element']
        
        # Create a grid of potential interstitial positions
        grid_density = max(5, int(n_configs ** (1/3)) + 2)
        grid_points = []
        
        for x in np.linspace(0.1, 0.9, grid_density):
            for y in np.linspace(0.1, 0.9, grid_density):
                for z in np.linspace(0.1, 0.9, grid_density):
                    grid_points.append([x, y, z])
        
        # Filter positions that are far enough from existing atoms
        valid_positions = []
        for frac_coords in grid_points:
            cart_coords = bulk.lattice.get_cartesian_coords(frac_coords)
            min_dist = min([np.linalg.norm(site.coords - cart_coords) for site in bulk])
            if min_dist > 1.5:  # Minimum 1.5 √Ö from any atom
                valid_positions.append(frac_coords)
        
        log(f"Found {len(valid_positions)} valid interstitial positions")
        
        # Sample positions
        n_sample = min(n_configs, len(valid_positions))
        if strategy == "random":
            sampled_positions = random.sample(valid_positions, n_sample)
        else:
            sampled_positions = valid_positions[:n_sample]
        
        # Create configurations
        for i, frac_coords in enumerate(sampled_positions):
            struct = bulk.copy()
            cart_coords = struct.lattice.get_cartesian_coords(frac_coords)
            struct.append(element, cart_coords, coords_are_cartesian=True)
            struct.perturb(0.05)
            configs.append((struct, f"{defect_str}_pos{i}"))
    
    elif defect_info['type'] == 'complex':
        # For complex defects, use the original create_defect_structure
        # but apply random perturbations to create different configurations
        log(f"Complex defect with {len(defect_info['parts'])} parts")
        
        for i in range(n_configs):
            try:
                struct = defect_db_create_defect_structure(bulk, defect_str)
                # Apply stronger random perturbation for variation
                struct.perturb(0.10 + 0.05 * np.random.random())
                configs.append((struct, f"{defect_str}_config{i}"))
            except Exception as e:
                log(f"  Failed to create config {i}: {e}")
                continue
    
    log(f"Generated {len(configs)} configurations")
    return configs

# =====================================================================================
# OPTIMIZATION ENGINE
# =====================================================================================

def defect_db_optimize_configurations(configs: list, model_type: str, charge: int, fmax: float, steps: int,
                           progress_cb=None, log=print) -> pd.DataFrame:
    """Optimize all configurations and return results sorted by energy"""
    
    model_map = defect_db_charge_model_names(model_type)
    model_name = model_map[charge]
    
    # Load model
    log(f"Loading model: {model_name}")
    relaxer = defect_db_load_relaxer(model_type, model_name, relax_cell=False, log=log)
    
    results = []
    total = len(configs)
    
    for i, (struct, label) in enumerate(configs):
        log(f"Optimizing {i+1}/{total}: {label}")
        
        try:
            res = relaxer.relax(struct.copy(), fmax=fmax, steps=steps)
            
            E_initial = float(res["trajectory"].energies[0])
            E_final = float(res["trajectory"].energies[-1])
            n_steps = len(res["trajectory"].energies)
            converged = n_steps < steps
            
            results.append({
                'config_id': i,
                'label': label,
                'E_initial': E_initial,
                'E_final': E_final,
                'E_diff': E_final - E_initial,
                'n_steps': n_steps,
                'converged': converged,
                'structure_cif': str(CifWriter(res["final_structure"]))
            })
            
            log(f"  ‚úì E_final = {E_final:.6f} eV ({n_steps} steps)")
            
        except Exception as e:
            log(f"  ‚úó Failed: {e}")
            results.append({
                'config_id': i,
                'label': label,
                'E_initial': np.nan,
                'E_final': np.nan,
                'E_diff': np.nan,
                'n_steps': 0,
                'converged': False,
                'structure_cif': None
            })
        
        if progress_cb:
            progress_cb(i+1, total)
    
    # Create DataFrame and sort by energy
    df = pd.DataFrame(results)
    df = df.sort_values('E_final').reset_index(drop=True)
    
    return df

# =====================================================================================
# VISUALIZATION
# =====================================================================================

def defect_db_plot_energy_distribution(df: pd.DataFrame, defect_label: str):
    """Plot energy distribution and landscape"""
    fig = plt.figure(figsize=(16, 6), facecolor='white')
    gs = GridSpec(1, 2, figure=fig)
    
    # Histogram
    ax1 = fig.add_subplot(gs[0, 0])
    valid_energies = df['E_final'].dropna()
    
    if len(valid_energies) > 0:
        ax1.hist(valid_energies, bins=min(30, len(valid_energies)//2), 
                color='#8b5cf6', alpha=0.7, edgecolor='black', linewidth=1.5)
        ax1.axvline(valid_energies.min(), color='#dc2626', linestyle='--', 
                   linewidth=2.5, label=f'Min: {valid_energies.min():.3f} eV')
        ax1.axvline(valid_energies.mean(), color='#10b981', linestyle='--', 
                   linewidth=2.5, label=f'Mean: {valid_energies.mean():.3f} eV')
        ax1.set_xlabel('Energy (eV)', fontsize=18, fontweight='bold')
        ax1.set_ylabel('Count', fontsize=18, fontweight='bold')
        ax1.set_title(f'Energy Distribution: {defect_label}', fontsize=20, fontweight='bold')
        ax1.legend(fontsize=16, framealpha=0.95)
        ax1.grid(True, alpha=0.3, linestyle='--')
        ax1.tick_params(labelsize=16)
    
    # Energy landscape
    ax2 = fig.add_subplot(gs[0, 1])
    valid_df = df.dropna(subset=['E_final'])
    
    if len(valid_df) > 0:
        colors = plt.cm.plasma(np.linspace(0, 1, len(valid_df)))
        ax2.scatter(range(len(valid_df)), valid_df['E_final'], 
                   c=range(len(valid_df)), cmap='plasma', s=80, 
                   alpha=0.7, edgecolors='black', linewidth=1)
        ax2.plot(range(len(valid_df)), valid_df['E_final'], 
                'k-', alpha=0.3, linewidth=1.5)
        
        # Highlight top 5
        for i in range(min(5, len(valid_df))):
            ax2.scatter(i, valid_df.iloc[i]['E_final'], 
                       s=200, facecolors='none', edgecolors='red', 
                       linewidths=2.5, zorder=10)
        
        ax2.set_xlabel('Configuration (sorted by energy)', fontsize=18, fontweight='bold')
        ax2.set_ylabel('Energy (eV)', fontsize=18, fontweight='bold')
        ax2.set_title('Energy Landscape (Red = Top 5)', fontsize=20, fontweight='bold')
        ax2.grid(True, alpha=0.3, linestyle='--')
        ax2.tick_params(labelsize=16)
    
    plt.tight_layout()
    return fig

def defect_db_create_download_link(content, filename, content_type='text/plain'):
    if isinstance(content, str):
        content = content.encode()
    
    b64 = base64.b64encode(content).decode()
    href = f'data:{content_type};base64,{b64}'
    
    html = f'''
    <a href="{href}" 
       download="{filename}" 
       style="display:none" 
       id="dl_{filename.replace(".", "_").replace(" ", "_")}"></a>
    <script>
        document.getElementById("dl_{filename.replace(".", "_").replace(" ", "_")}").click();
    </script>
    '''
    return HTML(html)

# =====================================================================================
# GUI - CREATE ALL WIDGETS
# =====================================================================================

defect_db_header = widgets.HTML("""
<style>
    .sampling-header {
        background: linear-gradient(135deg, #9ca3af 0%, #6b7280 100%);
        padding: 30px;
        border-radius: 15px;
        box-shadow: 0 10px 30px rgba(0,0,0,0.2);
        margin-bottom: 20px;
    }
    .sampling-title {
        color: white;
        font-size: 32px;
        font-weight: 800;
        margin: 0;
        text-shadow: 2px 2px 4px rgba(0,0,0,0.3);
        text-align: center;
    }
    .sampling-subtitle {
        color: #f3f4f6;
        font-size: 16px;
        margin-top: 8px;
        text-align: center;
    }
</style>
<div class="sampling-header">
    <div class="sampling-title">üéØ DeFecT-FF - Configuration Sampling</div>
    <div class="sampling-subtitle">
        Sample Multiple Defect Positions to Find Lowest Energy Configuration
    </div>
</div>
""")

# Composition selector
defect_db_compositions = [
    "Cd0.25Zn0.75Te", "Cd0.50Zn0.50Te", "Cd0.75Zn0.25Te",
    "CdSe", "CdSe0.06Te0.94", "CdSe0.12Te0.88",
    "CdSe0.20Te0.80", "CdSe0.25Te0.75", "CdSe0.50Te0.50",
    "CdSe0.75Te0.25", "CdTe", "ZnTe"
]

defect_db_composition_dropdown = widgets.Dropdown(
    options=["Select composition..."] + defect_db_compositions,
    value="Select composition...",
    description="Bulk:",
    style={'description_width': '100px'},
    layout=widgets.Layout(width='100%')
)

# Defect input
defect_db_defect_input = widgets.Text(
    value="V_Cd",
    placeholder="e.g., V_Cd, Cu_i, As_Te, Cl_Te+V_Cd",
    description="Defect:",
    style={'description_width': '100px'},
    layout=widgets.Layout(width='100%')
)

defect_db_syntax_card = widgets.HTML("""
<div style="padding:10px;border-radius:8px;background:#f5f7fa;
            border-left:4px solid #9ca3af;margin:5px 0;font-size:12px;">
  <b>Defect Syntax:</b> 
  <code>V_Cd</code> (vacancy) | <code>As_Te</code> (substitution) | 
  <code>Cu_i</code> (interstitial) | <code>Cl_Te+V_Cd</code> (complex)
</div>
""")

# Sampling parameters
defect_db_n_configs = widgets.IntSlider(
    value=100,
    min=10,
    max=500,
    step=10,
    description="N configs:",
    style={'description_width': '100px'},
    layout=widgets.Layout(width='100%')
)

defect_db_sampling_strategy = widgets.Dropdown(
    options=["random", "systematic", "all"],
    value="random",
    description="Strategy:",
    style={'description_width': '100px'},
    layout=widgets.Layout(width='100%')
)

defect_db_charge_state = widgets.Dropdown(
    options=[("Neutral (0)", 0), ("+1", 1), ("+2", 2), ("-1", -1), ("-2", -2)],
    value=0,
    description="Charge:",
    style={'description_width': '100px'},
    layout=widgets.Layout(width='100%')
)

# Model settings
defect_db_model_toggle = widgets.ToggleButtons(
    options=["PBE", "HSE"],
    value="PBE",
    description="Model:",
    style={"description_width": "80px"},
    layout=widgets.Layout(width='100%')
)

defect_db_steps_slider = widgets.IntSlider(
    value=200, min=50, max=1000, step=50,
    description="Steps:",
    style={"description_width": "80px"},
    layout=widgets.Layout(width='100%')
)

defect_db_fmax_slider = widgets.FloatSlider(
    value=0.05, min=0.01, max=0.2, step=0.01,
    description="Fmax:",
    readout_format=".2f",
    style={"description_width": "80px"},
    layout=widgets.Layout(width='100%')
)

# Buttons
defect_db_run_button = widgets.Button(
    description="üöÄ Run Sampling",
    button_style="",
    layout=widgets.Layout(width='100%', height='50px'),
    style={'font_weight': 'bold'}
)

defect_db_download_button = widgets.Button(
    description="üì• Download Results",
    button_style="",
    layout=widgets.Layout(width='100%', height='50px'),
    style={'font_weight': 'bold'},
    disabled=True
)

# Progress
defect_db_progress = widgets.IntProgress(
    value=0, min=0, max=1,
    description="Progress:",
    style={"description_width": "80px", 'bar_color': '#9ca3af'},
    layout=widgets.Layout(width='100%')
)

defect_db_status = widgets.HTML(
    "<div style='padding:8px;background:#f9fafb;border-left:4px solid #9ca3af;border-radius:6px;font-size:12px;'>"
    "<b>Status:</b> Ready</div>"
)

# Outputs
defect_db_log_output = widgets.Output(
    layout={"border": "2px solid #e2e8f0", "border_radius": "8px",
            "padding": "10px", "height": "350px", "overflow": "auto", "background": "#1a202c"}
)
defect_db_log_output._dom_classes = [f'{DEFECT_DB_APP_ID}_log']

defect_db_results_output = widgets.Output(
    layout={"border": "2px solid #e2e8f0", "border_radius": "8px", "padding": "10px"}
)
defect_db_results_output._dom_classes = [f'{DEFECT_DB_APP_ID}_results']

# Global variables
defect_db_global_results = None

# =====================================================================================
# EVENT HANDLERS
# =====================================================================================

def defect_db_set_progress(cur, total):
    defect_db_progress.max = max(1, int(total))
    defect_db_progress.value = min(int(cur), defect_db_progress.max)

def defect_db_on_run_clicked(b):
    global defect_db_global_results
    
    defect_db_log_output.clear_output(wait=True)
    with defect_db_log_output:
        
        # Validate inputs
        if defect_db_composition_dropdown.value == "Select composition...":
            print("\033[91m[ERROR] Please select a composition\033[0m")
            return
        
        defect_str = defect_db_defect_input.value.strip()
        if not defect_str:
            print("\033[91m[ERROR] Please enter a defect\033[0m")
            return
        
        composition = defect_db_composition_dropdown.value
        n_conf = defect_db_n_configs.value
        
        print(f"\033[96m=== DEFECT CONFIGURATION SAMPLING ===\033[0m")
        print(f"Composition: {composition}")
        print(f"Defect: {defect_str}")
        print(f"Defect Type: {defect_db_get_defect_type(defect_str)}")
        print(f"N Configurations: {n_conf}")
        print(f"Strategy: {defect_db_sampling_strategy.value}")
        print(f"Charge State: {defect_db_charge_state.value:+d}")
        print(f"Model: {defect_db_model_toggle.value}")
        print("=" * 80)
        
        defect_db_status.value = ("<div style='padding:8px;background:#f3f4f6;border-left:4px solid #9ca3af;"
                       "border-radius:6px;font-size:12px;'><b>Status:</b> Loading bulk...</div>")
        
        # Load bulk structure
        try:
            bulk_path = defect_db_fetch_bulk_from_github(composition, log=lambda msg: print(f"\033[93m[FETCH] {msg}\033[0m"))
            bulk = defect_db_read_structure(bulk_path)
            print(f"\033[92m[INFO] Loaded bulk: {len(bulk)} atoms\033[0m")
            print(f"  Species: {sorted(set(site.specie.symbol for site in bulk))}")
        except Exception as e:
            print(f"\033[91m[ERROR] Failed to load bulk: {e}\033[0m")
            return
        
        # Generate configurations
        defect_db_status.value = ("<div style='padding:8px;background:#f3f4f6;border-left:4px solid #9ca3af;"
                       "border-radius:6px;font-size:12px;'><b>Status:</b> Generating configurations...</div>")
        
        print(f"\033[96m\n=== CONFIGURATION GENERATION ===\033[0m")
        try:
            configs = defect_db_generate_defect_configurations(
                bulk, 
                defect_str, 
                n_conf, 
                defect_db_sampling_strategy.value,
                log=lambda msg: print(f"\033[93m[GEN] {msg}\033[0m")
            )
            
            print(f"\033[92m[SUCCESS] Generated {len(configs)} configurations\033[0m")
            
        except Exception as e:
            print(f"\033[91m[ERROR] Configuration generation failed: {e}\033[0m")
            import traceback
            traceback.print_exc()
            defect_db_status.value = ("<div style='padding:8px;background:#fee2e2;border-left:4px solid #9ca3af;"
                           "border-radius:6px;font-size:12px;'><b>Status:</b> Failed</div>")
            return
        
        # Optimize
        defect_db_status.value = ("<div style='padding:8px;background:#f3f4f6;border-left:4px solid #9ca3af;"
                       "border-radius:6px;font-size:12px;'><b>Status:</b> Optimizing...</div>")
        
        print(f"\033[96m\n=== MLFF OPTIMIZATION ===\033[0m")
        try:
            results_df = defect_db_optimize_configurations(
                configs,
                defect_db_model_toggle.value,
                defect_db_charge_state.value,
                defect_db_fmax_slider.value,
                defect_db_steps_slider.value,
                progress_cb=defect_db_set_progress,
                log=lambda msg: print(f"\033[93m[OPT] {msg}\033[0m")
            )
            
            print(f"\033[92m[SUCCESS] Optimization complete\033[0m")
            
        except Exception as e:
            print(f"\033[91m[ERROR] Optimization failed: {e}\033[0m")
            import traceback
            traceback.print_exc()
            defect_db_status.value = ("<div style='padding:8px;background:#fee2e2;border-left:4px solid #9ca3af;"
                           "border-radius:6px;font-size:12px;'><b>Status:</b> Failed</div>")
            return
        
        # Analysis
        print(f"\033[96m\n=== RESULTS SUMMARY ===\033[0m")
        valid_results = results_df.dropna(subset=['E_final'])
        
        if len(valid_results) > 0:
            print(f"Successfully optimized: {len(valid_results)}/{len(results_df)}")
            print(f"\nEnergy Statistics:")
            print(f"  Lowest:  {valid_results['E_final'].min():.6f} eV")
            print(f"  Highest: {valid_results['E_final'].max():.6f} eV")
            print(f"  Mean:    {valid_results['E_final'].mean():.6f} eV")
            print(f"  Std Dev: {valid_results['E_final'].std():.6f} eV")
            print(f"  Range:   {valid_results['E_final'].max() - valid_results['E_final'].min():.6f} eV")
            
            print(f"\nüèÜ Top 10 Lowest Energy Configurations:")
            for idx, row in valid_results.head(10).iterrows():
                print(f"  {idx+1:2d}. {row['label']:30s} ‚Üí {row['E_final']:.6f} eV")
        else:
            print("\033[91m[ERROR] No successful optimizations\033[0m")
            defect_db_status.value = ("<div style='padding:8px;background:#fee2e2;border-left:4px solid #9ca3af;"
                           "border-radius:6px;font-size:12px;'><b>Status:</b> No results</div>")
            return
        
        # Plot
        print(f"\033[96m\n=== GENERATING PLOTS ===\033[0m")
    
    # PLOT OUTSIDE log_output context to ensure it displays in results_output
    defect_db_results_output.clear_output(wait=True)
    with defect_db_results_output:
        try:
            fig = defect_db_plot_energy_distribution(results_df, defect_str)
            plt.show()
        except Exception as e:
            print(f"Plot error: {e}")
    
    # Continue in log for final messages
    with defect_db_log_output:
        # Store results
        defect_db_global_results = {
            'df': results_df,
            'composition': composition,
            'defect': defect_str,
            'charge': defect_db_charge_state.value,
            'model': defect_db_model_toggle.value,
            'figure': fig
        }
        
        defect_db_download_button.disabled = False
        defect_db_status.value = ("<div style='padding:8px;background:#f9fafb;border-left:4px solid #9ca3af;"
                       "border-radius:6px;font-size:12px;'><b>Status:</b> Complete! ‚úì</div>")
        print("\n" + "=" * 80)
        print("\033[92m[COMPLETE] Configuration sampling finished!\033[0m")

defect_db_run_button.on_click(defect_db_on_run_clicked)

def defect_db_on_download_clicked(b):
    global defect_db_global_results
    
    if defect_db_global_results is None:
        return
    
    df = defect_db_global_results['df']
    valid_df = df.dropna(subset=['E_final', 'structure_cif'])
    
    if len(valid_df) == 0:
        return
    
    zip_buffer = io.BytesIO()
    
    with zipfile.ZipFile(zip_buffer, 'w', zipfile.ZIP_DEFLATED) as zf:
        # Energy table
        zf.writestr("energy_table.csv", df.to_csv(index=False))
        
        # Lowest energy structure
        lowest_row = valid_df.iloc[0]
        zf.writestr(f"LOWEST_ENERGY_{lowest_row['label']}.cif", lowest_row['structure_cif'])
        
        # Top 10 structures
        for idx, row in valid_df.head(10).iterrows():
            zf.writestr(f"top10/rank{idx+1:02d}_{row['label']}.cif", row['structure_cif'])
        
        # All structures
        for idx, row in valid_df.iterrows():
            if pd.notna(row['structure_cif']):
                zf.writestr(f"all_structures/{row['label']}.cif", row['structure_cif'])
        
        # Plot
        img_buffer = io.BytesIO()
        defect_db_global_results['figure'].savefig(img_buffer, format='png', dpi=300, bbox_inches='tight')
        img_buffer.seek(0)
        zf.writestr("energy_distribution.png", img_buffer.read())
        
        # README
        readme = f"""
# DefectDB - Configuration Sampling Results

Composition: {defect_db_global_results['composition']}
Defect: {defect_db_global_results['defect']}
Defect Type: {defect_db_get_defect_type(defect_db_global_results['defect'])}
Charge State: {defect_db_global_results['charge']:+d}
Model: {defect_db_global_results['model']}
Date: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}

## Summary Statistics
Total Configurations: {len(df)}
Successful Optimizations: {len(valid_df)}
Lowest Energy: {valid_df['E_final'].min():.6f} eV
Highest Energy: {valid_df['E_final'].max():.6f} eV
Mean Energy: {valid_df['E_final'].mean():.6f} eV
Std Deviation: {valid_df['E_final'].std():.6f} eV
Energy Range: {valid_df['E_final'].max() - valid_df['E_final'].min():.6f} eV

## Files
- energy_table.csv: Full results table with all configurations
- LOWEST_ENERGY_*.cif: The lowest energy configuration (MOST IMPORTANT)
- top10/: Top 10 lowest energy structures
- all_structures/: All optimized structures
- energy_distribution.png: Energy distribution and landscape plots

## Interpretation
The LOWEST_ENERGY structure represents the most stable defect configuration
at the specified charge state. Use this structure for further analysis or
as the initial structure for more refined calculations.
"""
        zf.writestr("README.txt", readme)
    
    zip_buffer.seek(0)
    display(defect_db_create_download_link(
        zip_buffer.read(),
        f"{defect_db_global_results['composition']}_{defect_db_global_results['defect']}_q{defect_db_global_results['charge']:+d}_SAMPLING.zip",
        "application/zip"
    ))

defect_db_download_button.on_click(defect_db_on_download_clicked)

# =====================================================================================
# LAYOUT - BUILD INTERFACE
# =====================================================================================

defect_db_instructions = widgets.HTML("""
<div style="background:#f5f7fa;padding:15px;border-radius:8px;border-left:4px solid #9ca3af;margin:10px 0;font-size:13px;">
    <b style="color:#4b5563;">üìö How This Tool Works:</b>
    <ol style="margin:5px 0;padding-left:20px;color:#6b7280;line-height:1.8;">
        <li>Select bulk composition from database</li>
        <li>Define defect using syntax: <code>V_Cd</code>, <code>Cu_i</code>, <code>As_Te</code>, <code>Cl_Te+V_Cd</code></li>
        <li>Tool generates N configurations at different atomic positions</li>
        <li>Each configuration is optimized with MLFF</li>
        <li>Results sorted by energy to identify lowest energy structure</li>
        <li>Download lowest energy + top 10 + all structures</li>
    </ol>
</div>
""")

defect_db_left_panel = widgets.VBox([
    widgets.HTML("<h4 style='color:#4b5563;font-weight:700;margin:0 0 5px 0;'>Step 1: Select Bulk</h4>"),
    defect_db_composition_dropdown,
    widgets.HTML("<h4 style='color:#4b5563;font-weight:700;margin:15px 0 5px 0;'>Step 2: Define Defect</h4>"),
    defect_db_defect_input,
    defect_db_syntax_card,
    widgets.HTML("<h4 style='color:#4b5563;font-weight:700;margin:15px 0 5px 0;'>Step 3: Sampling Parameters</h4>"),
    defect_db_n_configs,
    defect_db_sampling_strategy,
    defect_db_charge_state,
    widgets.HTML("<h4 style='color:#4b5563;font-weight:700;margin:15px 0 5px 0;'>Step 4: Optimization Settings</h4>"),
    defect_db_model_toggle,
    defect_db_steps_slider,
    defect_db_fmax_slider,
    widgets.HTML("<h4 style='color:#4b5563;font-weight:700;margin:15px 0 5px 0;'>Step 5: Run</h4>"),
    defect_db_run_button,
    defect_db_progress,
    defect_db_status,
    defect_db_download_button,
], layout=widgets.Layout(width='48%', padding='10px'))

defect_db_right_panel = widgets.VBox([
    widgets.HTML("<h4 style='color:#4b5563;font-weight:700;margin:0 0 5px 0;'>üìü Execution Log</h4>"),
    defect_db_log_output,
    widgets.HTML("<h4 style='color:#4b5563;font-weight:700;margin:15px 0 5px 0;'>üìä Results</h4>"),
    defect_db_results_output,
], layout=widgets.Layout(width='48%', padding='10px'))

defect_db_main_layout = widgets.VBox([
    defect_db_header,
    defect_db_instructions,
    widgets.HBox([defect_db_left_panel, defect_db_right_panel], layout=widgets.Layout(justify_content='space-between'))
])


display(defect_db_main_layout)
```

```python
# =====================================================================================
# üß™ Chemical Potential Calculator - Standalone Tool
# All variables prefixed with chempot_ to avoid conflicts
# =====================================================================================

import json, requests, warnings, os, zipfile, time, io, base64, tempfile, shutil, re
from pathlib import Path
from itertools import permutations
from datetime import datetime

import numpy as np
import pandas as pd

import ipywidgets as widgets
from IPython.display import display, clear_output, HTML

from pymatgen.core import Structure, Composition
from pymatgen.core.periodic_table import Element
from pymatgen.io.vasp import Vasprun
from pymatgen.entries.computed_entries import ComputedStructureEntry

try:
    from doped.chemical_potentials import CompetingPhasesAnalyzer
    chempot_HAS_DOPED = True
except ImportError:
    chempot_HAS_DOPED = False
    print("‚ö†Ô∏è ERROR: 'doped' package required. Install: pip install doped")

# =====================================================================================
# SUPPRESS WARNINGS
# =====================================================================================
warnings.filterwarnings("ignore")
import logging
logging.getLogger('doped').setLevel(logging.ERROR)

# =====================================================================================
# CONFIGURATION
# =====================================================================================
chempot_COMPETING_CACHE_DIR = Path.home() / "chempot_competing_phases_cache"
chempot_COMPETING_CACHE_DIR.mkdir(parents=True, exist_ok=True)

chempot_GITHUB_RELEASE_API = "https://api.github.com/repos/msehabibur/nanohub-simple-jupyter-tool/releases/tags/CompetingPhases"
chempot_ANIONS = {"O", "S", "Se", "Te"}

# =====================================================================================
# COMPOSITION MATCHING
# =====================================================================================

def chempot_find_best_matching_composition(target_comp: Composition, entries: list, log=print) -> str:
    """
    Find the best matching composition from entries.
    Compares reduced compositions (element ratios).
    
    Example: Cd25Se3Te22 ‚Üí matches Cd9Te8Se (both have same element ratios)
    """
    target_reduced = target_comp.reduced_composition
    target_elements = set(str(el) for el in target_reduced.elements)
    
    log(f"üîç Searching for composition matching: {target_reduced.formula}")
    log(f"   Target elements: {sorted(target_elements)}")
    
    # Get all unique compositions from entries
    entry_compositions = {}
    for entry in entries:
        comp = entry.composition.reduced_composition
        comp_elements = set(str(el) for el in comp.elements)
        
        # Only consider entries with same elements
        if comp_elements == target_elements:
            formula = comp.formula
            entry_compositions[formula] = comp
    
    log(f"   Found {len(entry_compositions)} phases with same elements")
    
    if not entry_compositions:
        available = set()
        for entry in entries:
            comp_els = set(str(el) for el in entry.composition.elements)
            if comp_els.issubset(target_elements):
                available.add(entry.composition.reduced_formula)
        raise ValueError(f"No matching composition found. Available phases: {sorted(available)}")
    
    # Try exact match first
    if target_reduced.formula in entry_compositions:
        log(f"‚úÖ Found exact match: {target_reduced.formula}")
        return target_reduced.formula
    
    # Find closest match by comparing fractional amounts
    best_match = None
    best_distance = float('inf')
    
    for formula, comp in entry_compositions.items():
        distance = 0.0
        for el in target_elements:
            target_frac = target_reduced.get_atomic_fraction(el)
            comp_frac = comp.get_atomic_fraction(el)
            distance += abs(target_frac - comp_frac)
        
        log(f"   {formula}: distance = {distance:.6f}")
        
        if distance < best_distance:
            best_distance = distance
            best_match = formula
    
    if best_match and best_distance < 0.01:  # Within 1% tolerance
        log(f"‚úÖ Found close match: {best_match} (distance: {best_distance:.6f})")
        return best_match
    
    log(f"‚ö†Ô∏è Using best available match: {best_match} (distance: {best_distance:.6f})")
    return best_match

# =====================================================================================
# COMPETING PHASES DOWNLOAD + SMART FILTERING
# =====================================================================================

def chempot_extract_elements_from_folder_name(folder_name: str) -> set:
    """Extract element symbols from folder name like 'CdSe' ‚Üí {'Cd', 'Se'}"""
    pattern = r'([A-Z][a-z]?)'
    elements = re.findall(pattern, folder_name)
    return set(elements)

def chempot_filter_competing_phases_by_elements(extract_dir: Path, required_elements: set, log=print):
    """
    CRITICAL: Keep only vasprun.xml files relevant to required_elements.
    Delete everything else to avoid OOM.
    """
    log(f"üîç Filtering CompetingPhases for elements: {sorted(required_elements)}")
    
    kept_files = []
    deleted_files = []
    
    # Scan all vasprun files
    for vasprun_path in extract_dir.rglob("vasprun.xml*"):
        if not vasprun_path.is_file():
            continue
        
        # Extract elements from parent folder name
        folder_name = vasprun_path.parent.name
        folder_elements = chempot_extract_elements_from_folder_name(folder_name)
        
        # Keep if folder elements are subset of required elements
        if folder_elements and folder_elements.issubset(required_elements):
            kept_files.append(vasprun_path)
        else:
            # Delete irrelevant file
            try:
                vasprun_path.unlink()
                deleted_files.append(vasprun_path)
            except Exception:
                pass
    
    # Delete empty directories
    for dirpath in list(extract_dir.rglob("*")):
        if dirpath.is_dir():
            try:
                # Try to remove if empty
                dirpath.rmdir()
            except OSError:
                # Directory not empty, skip
                pass
    
    log(f"‚úÖ Kept {len(kept_files)} relevant vasprun files")
    log(f"üóëÔ∏è Deleted {len(deleted_files)} irrelevant files")
    
    return kept_files

def chempot_download_and_filter_competing_phases(composition: str, impurities: list, force_redownload=False, log=print):
    """
    Download CompetingPhases.zip and IMMEDIATELY filter by composition + impurities
    """
    zip_path = chempot_COMPETING_CACHE_DIR / "CompetingPhases.zip"
    extract_dir = chempot_COMPETING_CACHE_DIR / "CompetingPhases"
    
    # Determine required elements
    comp_obj = Composition(composition)
    host_elements = set(str(el) for el in comp_obj.elements)
    required_elements = host_elements | set(impurities)
    
    log(f"Required elements: {sorted(required_elements)}")
    
    # Check if already filtered for these elements
    marker_file = extract_dir / f".filtered_{'-'.join(sorted(required_elements))}"
    
    if marker_file.exists() and not force_redownload:
        log(f"‚úÖ Using cached filtered CompetingPhases")
        vasprun_files = list(extract_dir.rglob("vasprun.xml*"))
        log(f"   Found {len(vasprun_files)} vasprun files")
        return extract_dir
    
    # Download
    log("üì• Downloading CompetingPhases.zip from GitHub...")
    log("   This may take a few minutes (large file ~50 MB)...")
    
    try:
        # Get release info
        response = requests.get(chempot_GITHUB_RELEASE_API, timeout=60)
        response.raise_for_status()
        release_data = response.json()
        
        # Find asset
        asset_url = None
        for asset in release_data.get('assets', []):
            if 'CompetingPhases' in asset.get('name', '') and asset['name'].endswith('.zip'):
                asset_url = asset['browser_download_url']
                break
        
        if not asset_url:
            raise Exception("CompetingPhases.zip not found in release")
        
        # Download with progress
        response = requests.get(asset_url, stream=True, timeout=300)
        response.raise_for_status()
        
        total_size = int(response.headers.get('content-length', 0))
        downloaded = 0
        
        with open(zip_path, 'wb') as f:
            for chunk in response.iter_content(chunk_size=8192):
                if chunk:
                    f.write(chunk)
                    downloaded += len(chunk)
                    if total_size > 0:
                        percent = (downloaded / total_size) * 100
                        if downloaded % (5 * 1024 * 1024) == 0:  # Log every 5 MB
                            log(f"   Progress: {percent:.1f}%")
        
        log(f"‚úÖ Download complete")
        
        # Extract
        log("üìÇ Extracting ZIP...")
        if extract_dir.exists():
            shutil.rmtree(extract_dir)
        extract_dir.mkdir(parents=True, exist_ok=True)
        
        with zipfile.ZipFile(zip_path, 'r') as zf:
            zf.extractall(extract_dir)
        
        log(f"‚úÖ Extracted to: {extract_dir}")
        
        # IMMEDIATE FILTERING
        log("\nüîß FILTERING: Keeping only relevant files...")
        kept_files = chempot_filter_competing_phases_by_elements(extract_dir, required_elements, log=log)
        
        # Create marker file
        marker_file.write_text(f"Filtered for: {sorted(required_elements)}")
        
        if len(kept_files) == 0:
            raise Exception(f"No relevant vasprun files found for {required_elements}")
        
        return extract_dir
        
    except Exception as e:
        raise Exception(f"Failed to download/filter competing phases: {e}")

# =====================================================================================
# CANONICAL FORMULA
# =====================================================================================

def chempot_canonical_formula(composition: str) -> str:
    """Convert composition to canonical formula (anions last, hide 1s)"""
    try:
        comp = Composition(composition)
        integer_form, _ = comp.get_integer_formula_and_factor()
        
        if isinstance(integer_form, str):
            comp_int = Composition(integer_form)
        else:
            comp_int = comp
        
        elements = sorted(comp_int.elements, 
                         key=lambda el: (1 if str(el) in chempot_ANIONS else 0, el.Z))
        
        parts = []
        for el in elements:
            amt = int(round(comp_int[el]))
            parts.append(f"{el}{'' if amt == 1 else amt}")
        
        return "".join(parts)
    except Exception:
        return composition

# =====================================================================================
# CHEMICAL POTENTIAL CALCULATION
# =====================================================================================

def chempot_find_matching_vasprun_files(composition: str, elements: list, extract_dir: Path, log=print) -> list:
    """Find relevant vasprun files for given elements"""
    element_set = set(elements)
    valid_files = []
    
    for vr_path in extract_dir.rglob("vasprun.xml*"):
        if not vr_path.is_file():
            continue
        
        folder_elements = chempot_extract_elements_from_folder_name(vr_path.parent.name)
        
        if folder_elements and folder_elements.issubset(element_set):
            valid_files.append(str(vr_path))
    
    log(f"Found {len(valid_files)} relevant vasprun files")
    return valid_files

def chempot_extract_elemental_references(vasprun_files: list, elements: list, log=print) -> dict:
    """Extract elemental reference energies Œº¬∞_el (eV/atom)"""
    refs = {}
    
    for vr_path in vasprun_files:
        try:
            vr = Vasprun(vr_path, parse_dos=False, parse_eigen=False,
                        parse_potcar_file=False, exception_on_bad_xml=False)
            comp = vr.final_structure.composition
            comp_elements = set(str(el) for el in comp.elements)
            
            # Elemental phase (single element)
            if len(comp_elements) == 1:
                el = list(comp_elements)[0]
                if el in elements:
                    refs[el] = vr.final_energy / vr.final_structure.num_sites
                    log(f"  Reference E({el}) = {refs[el]:.6f} eV/atom")
        except Exception:
            continue
    
    return refs

def chempot_vasprun_to_entry(vr_path: str) -> ComputedStructureEntry:
    """Convert vasprun to ComputedStructureEntry"""
    try:
        vr = Vasprun(vr_path, parse_dos=False, parse_eigen=False,
                    parse_potcar_file=False, exception_on_bad_xml=False)
        return vr.get_computed_entry(inc_structure=True)
    except Exception:
        return None

def chempot_calculate_chemical_potentials_native(composition: str, extract_dir: Path, log=print):
    """Calculate chemical potentials for NATIVE defects only (FIXED COMPOSITION MATCHING)"""
    canonical = chempot_canonical_formula(composition)
    comp_obj = Composition(composition)
    host_elements = [str(el) for el in comp_obj.elements]
    
    log(f"\n{'='*80}")
    log(f"NATIVE DEFECTS")
    log(f"Composition: {composition} ‚Üí {canonical}")
    log(f"Elements: {host_elements}")
    log(f"{'='*80}")
    
    # Find vasprun files
    vasprun_files = chempot_find_matching_vasprun_files(composition, host_elements, extract_dir, log=log)
    
    if len(vasprun_files) == 0:
        raise RuntimeError(f"No vasprun files found for {host_elements}")
    
    # Extract elemental references
    log("\nExtracting elemental references...")
    elemental_refs = chempot_extract_elemental_references(vasprun_files, host_elements, log=log)
    
    # Build entries
    log("\nBuilding ComputedStructureEntry objects...")
    entries = []
    for vr_path in vasprun_files:
        entry = chempot_vasprun_to_entry(vr_path)
        if entry:
            entries.append(entry)
    
    log(f"Created {len(entries)} entries")
    
    # ‚úÖ FIX: Find best matching composition from entries
    log("\nüîß Finding best matching composition from available phases...")
    try:
        target_comp = Composition(canonical)
        matched_formula = chempot_find_best_matching_composition(target_comp, entries, log=log)
    except Exception as e:
        log(f"‚ùå Composition matching failed: {e}")
        raise
    
    # Run CompetingPhasesAnalyzer with MATCHED composition
    log(f"\nRunning CompetingPhasesAnalyzer for {matched_formula}...")
    try:
        cpa = CompetingPhasesAnalyzer(matched_formula, entries)
        delta_mu_df = cpa.calculate_chempots()
    except Exception as e:
        log(f"‚ùå CompetingPhasesAnalyzer failed: {e}")
        raise
    
    # Calculate absolute mu
    absolute_mu_df = delta_mu_df.copy()
    
    if 'Limit' not in absolute_mu_df.columns:
        if absolute_mu_df.index.name == 'Limit':
            absolute_mu_df = absolute_mu_df.reset_index()
        else:
            absolute_mu_df['Limit'] = [f"Condition_{i+1}" for i in range(len(absolute_mu_df))]
    
    for el in host_elements:
        if el in delta_mu_df.columns and el in elemental_refs:
            absolute_mu_df[f"mu_{el} (eV)"] = delta_mu_df[el].values + elemental_refs[el]
    
    log("\n‚úÖ Native chemical potentials calculated")
    
    return {
        'delta_mu_df': delta_mu_df,
        'absolute_mu_df': absolute_mu_df,
        'elemental_refs': elemental_refs,
        'canonical': canonical,
        'matched_formula': matched_formula,
        'host_elements': host_elements
    }

def chempot_calculate_chemical_potentials_impurity(composition: str, impurity: str, extract_dir: Path, 
                                          native_refs: dict, log=print):
    """Calculate chemical potentials for ONE impurity (FIXED COMPOSITION MATCHING)"""
    canonical = chempot_canonical_formula(composition)
    comp_obj = Composition(composition)
    host_elements = [str(el) for el in comp_obj.elements]
    elements_with_imp = host_elements + [impurity]
    
    log(f"\n{'='*80}")
    log(f"IMPURITY: {impurity}")
    log(f"Elements: {elements_with_imp}")
    log(f"{'='*80}")
    
    # Find vasprun files
    vasprun_files = chempot_find_matching_vasprun_files(composition, elements_with_imp, extract_dir, log=log)
    
    if len(vasprun_files) == 0:
        log(f"‚ö†Ô∏è No vasprun files found for {impurity}, skipping...")
        return None
    
    # Extract elemental references (include impurity)
    log("\nExtracting elemental references...")
    elemental_refs = chempot_extract_elemental_references(vasprun_files, elements_with_imp, log=log)
    
    # Verify impurity reference exists
    if impurity not in elemental_refs:
        log(f"‚ö†Ô∏è No elemental reference found for {impurity}, skipping...")
        return None
    
    # Build entries
    log("\nBuilding ComputedStructureEntry objects...")
    entries = []
    for vr_path in vasprun_files:
        entry = chempot_vasprun_to_entry(vr_path)
        if entry:
            entries.append(entry)
    
    log(f"Created {len(entries)} entries")
    
    # ‚úÖ FIX: Find best matching composition
    log("\nüîß Finding best matching composition from available phases...")
    try:
        target_comp = Composition(canonical)
        matched_formula = chempot_find_best_matching_composition(target_comp, entries, log=log)
    except Exception as e:
        log(f"‚ùå Composition matching failed: {e}")
        return None
    
    # Run CompetingPhasesAnalyzer with MATCHED composition
    log(f"\nRunning CompetingPhasesAnalyzer for {matched_formula} with {impurity}...")
    try:
        cpa = CompetingPhasesAnalyzer(matched_formula, entries)
        delta_mu_df = cpa.calculate_chempots()
    except Exception as e:
        log(f"‚ùå Failed: {e}")
        return None
    
    # Calculate absolute mu
    absolute_mu_df = delta_mu_df.copy()
    
    if 'Limit' not in absolute_mu_df.columns:
        if absolute_mu_df.index.name == 'Limit':
            absolute_mu_df = absolute_mu_df.reset_index()
        else:
            absolute_mu_df['Limit'] = [f"Condition_{i+1}" for i in range(len(absolute_mu_df))]
    
    for el in host_elements:
        if el in delta_mu_df.columns and el in native_refs:
            absolute_mu_df[f"mu_{el} (eV)"] = delta_mu_df[el].values + native_refs[el]
    
    if impurity in delta_mu_df.columns:
        absolute_mu_df[f"mu_{impurity} (eV)"] = delta_mu_df[impurity].values + elemental_refs[impurity]
    
    log(f"\n‚úÖ Chemical potentials calculated for {impurity}")
    
    return {
        'delta_mu_df': delta_mu_df,
        'absolute_mu_df': absolute_mu_df,
        'elemental_refs': elemental_refs,
        'matched_formula': matched_formula,
        'impurity': impurity
    }

def chempot_generate_native_defects(elements: list) -> list:
    """Generate list of native defects: vacancies, antisites, interstitials"""
    defects = []
    defects += [f"V_{el}" for el in elements]
    for a, b in permutations(elements, 2):
        defects.append(f"{a}_{b}")
    defects += [f"{el}_i" for el in elements]
    return sorted(defects)

def chempot_calculate_defect_mu_table(absolute_mu_df: pd.DataFrame, host_elements: list, 
                             impurity: str = None) -> pd.DataFrame:
    """
    Calculate Œº(defect) for each condition:
    - V_X: Œº = Œº_X
    - X_i: Œº = -Œº_X
    - X_Y: Œº = Œº_Y - Œº_X
    """
    defects = chempot_generate_native_defects(host_elements)
    
    if impurity:
        defects.append(f"{impurity}_i")
        for host in host_elements:
            defects.append(f"{impurity}_{host}")
    
    defects = sorted(set(defects))
    
    rows = []
    for idx, row in absolute_mu_df.iterrows():
        condition = row['Limit']
        
        def get_mu(el: str) -> float:
            col = f"mu_{el} (eV)"
            return float(row[col]) if col in row else np.nan
        
        for defect in defects:
            if defect.startswith("V_"):
                X = defect.split("_")[1]
                mu_val = get_mu(X)
            elif defect.endswith("_i"):
                X = defect.split("_")[0]
                mu_val = -get_mu(X)
            else:
                A, B = defect.split("_")
                mu_val = get_mu(B) - get_mu(A)
            
            rows.append({
                'Defect': defect,
                'Condition': condition,
                'mu(defect) [eV]': mu_val
            })
    
    long_df = pd.DataFrame(rows)
    wide_df = long_df.pivot(index='Defect', columns='Condition', values='mu(defect) [eV]')
    wide_df = wide_df.reset_index()
    wide_df.columns.name = None
    
    return wide_df

# =====================================================================================
# DOWNLOAD UTILITIES
# =====================================================================================

def chempot_create_download_link(content, filename, content_type='text/plain'):
    if isinstance(content, str):
        content = content.encode()
    
    b64 = base64.b64encode(content).decode()
    href = f'data:{content_type};base64,{b64}'
    
    html = f'''
    <a href="{href}" 
       download="{filename}" 
       style="display:none" 
       id="dl_{filename.replace(".", "_").replace(" ", "_")}"></a>
    <script>
        document.getElementById("dl_{filename.replace(".", "_").replace(" ", "_")}").click();
    </script>
    '''
    return HTML(html)

# =====================================================================================
# GUI IMPLEMENTATION
# =====================================================================================

# Header
chempot_header = widgets.HTML("""
<style>
    .chempot-header {
        background: linear-gradient(135deg, #9ca3af 0%, #6b7280 100%);
        padding: 30px;
        border-radius: 15px;
        box-shadow: 0 10px 40px rgba(0,0,0,0.2);
        margin-bottom: 20px;
    }
    .chempot-title {
        color: white;
        font-size: 32px;
        font-weight: 800;
        margin: 0;
        text-shadow: 2px 2px 4px rgba(0,0,0,0.3);
        text-align: center;
    }
    .chempot-subtitle {
        color: #f3f4f6;
        font-size: 16px;
        margin-top: 10px;
        text-align: center;
    }
</style>
<div class="chempot-header">
    <div class="chempot-title">‚öóÔ∏è Chemical Potential Calculator</div>
    <div class="chempot-subtitle">
        Compute Œº for Native Defects + Impurities
    </div>
</div>
""")

# Composition selector
chempot_compositions = [
    "Cd0.25Zn0.75Te", "Cd0.50Zn0.50Te", "Cd0.75Zn0.25Te",
    "CdSe", "CdSe0.06Te0.94", "CdSe0.12Te0.88",
    "CdSe0.20Te0.80", "CdSe0.25Te0.75", "CdSe0.50Te0.50",
    "CdSe0.75Te0.25", "CdTe", "ZnTe"
]

chempot_composition_dropdown = widgets.Dropdown(
    options=["Select composition..."] + chempot_compositions,
    value="Select composition...",
    description="Composition:",
    style={'description_width': '120px'},
    layout=widgets.Layout(width='100%')
)

# Impurities input
chempot_impurities_text = widgets.Text(
    value="Cu,As,Cl",
    placeholder="e.g., Cu,As,Cl (comma-separated)",
    description="Impurities:",
    style={'description_width': '120px'},
    layout=widgets.Layout(width='100%')
)

# Run button
chempot_run_button = widgets.Button(
    description="üöÄ Calculate Chemical Potentials",
    button_style="",
    layout=widgets.Layout(width='100%', height='60px'),
    style={'font_weight': 'bold'}
)

# Download buttons
chempot_download_native_btn = widgets.Button(
    description="üìä Download Native Results",
    button_style='',
    layout=widgets.Layout(width='100%', height='50px'),
    disabled=True
)

chempot_download_impurities_btn = widgets.Button(
    description="üì¶ Download Impurity Results",
    button_style='',
    layout=widgets.Layout(width='100%', height='50px'),
    disabled=True
)

chempot_download_all_btn = widgets.Button(
    description="üìÅ Download ALL Results (ZIP)",
    button_style='',
    layout=widgets.Layout(width='100%', height='60px'),
    style={'font_weight': 'bold'},
    disabled=True
)

# Progress
chempot_progress = widgets.IntProgress(
    value=0, min=0, max=1,
    description="Progress:",
    style={"description_width": "80px", 'bar_color': '#9ca3af'},
    layout=widgets.Layout(width='100%')
)

chempot_status = widgets.HTML(
    "<div style='padding:10px;background:#f9fafb;border-left:4px solid #9ca3af;border-radius:8px;font-size:13px;'>"
    "<b>Status:</b> Ready</div>"
)

# Output area
chempot_log_output = widgets.Output(
    layout={"border": "2px solid #e2e8f0", "border_radius": "8px",
            "padding": "10px", "height": "400px", "overflow": "auto", "background": "#1a202c"}
)

# Global storage
chempot_global_results = None

# =====================================================================================
# EVENT HANDLERS
# =====================================================================================

def chempot_on_run_clicked(b):
    global chempot_global_results
    
    chempot_log_output.clear_output(wait=True)
    
    with chempot_log_output:
        # Validate inputs
        if chempot_composition_dropdown.value == "Select composition...":
            print("\033[91m[ERROR] Please select a composition\033[0m")
            return
        
        composition = chempot_composition_dropdown.value
        impurities_raw = chempot_impurities_text.value.strip()
        impurities = [x.strip() for x in impurities_raw.split(",") if x.strip()] if impurities_raw else []
        
        print(f"\033[96m=== CHEMICAL POTENTIAL CALCULATION ===\033[0m")
        print(f"Composition: {composition}")
        print(f"Impurities: {impurities if impurities else 'None (native only)'}")
        print("=" * 80)
        
        chempot_status.value = ("<div style='padding:10px;background:#f3f4f6;border-left:4px solid #9ca3af;"
                               "border-radius:8px;font-size:13px;'><b>Status:</b> Downloading CompetingPhases...</div>")
        
        # Download and filter CompetingPhases
        try:
            extract_dir = chempot_download_and_filter_competing_phases(
                composition,
                impurities,
                force_redownload=False,
                log=lambda msg: print(f"\033[93m[DOWNLOAD] {msg}\033[0m")
            )
        except Exception as e:
            print(f"\033[91m[ERROR] Failed to download/filter CompetingPhases: {e}\033[0m")
            chempot_status.value = ("<div style='padding:10px;background:#f3f4f6;border-left:4px solid #9ca3af;"
                                   "border-radius:8px;font-size:13px;'><b>Status:</b> Failed</div>")
            return
        
        chempot_status.value = ("<div style='padding:10px;background:#f3f4f6;border-left:4px solid #9ca3af;"
                               "border-radius:8px;font-size:13px;'><b>Status:</b> Calculating Œº (native)...</div>")
        
        # Calculate NATIVE
        try:
            native_results = chempot_calculate_chemical_potentials_native(
                composition,
                extract_dir,
                log=lambda msg: print(f"\033[93m[NATIVE] {msg}\033[0m")
            )
        except Exception as e:
            print(f"\033[91m[ERROR] Native calculation failed: {e}\033[0m")
            import traceback
            traceback.print_exc()
            chempot_status.value = ("<div style='padding:10px;background:#f3f4f6;border-left:4px solid #9ca3af;"
                                   "border-radius:8px;font-size:13px;'><b>Status:</b> Failed</div>")
            return
        
        # Calculate Œº(defect) for native
        native_defect_mu = chempot_calculate_defect_mu_table(
            native_results['absolute_mu_df'],
            native_results['host_elements']
        )
        native_results['defect_mu_table'] = native_defect_mu
        
        # Display native results
        print("\n" + "-"*80)
        print("NATIVE: Absolute Œº (eV)")
        print(f"Matched formula: {native_results['matched_formula']}")
        print("-"*80)
        display_cols = ['Limit'] + [col for col in native_results['absolute_mu_df'].columns if col.startswith('mu_')]
        print(native_results['absolute_mu_df'][display_cols].to_string(index=False))
        
        print("\n" + "-"*80)
        print("NATIVE: Œº(defect) [eV]")
        print("-"*80)
        print(native_defect_mu.to_string(index=False))
        
        # Store results
        results_by_impurity = {'native': native_results}
        
        # Calculate IMPURITIES
        for i, impurity in enumerate(impurities):
            chempot_status.value = (f"<div style='padding:10px;background:#f3f4f6;border-left:4px solid #9ca3af;"
                                   f"border-radius:8px;font-size:13px;'><b>Status:</b> Calculating Œº ({impurity})...</div>")
            
            try:
                imp_results = chempot_calculate_chemical_potentials_impurity(
                    composition,
                    impurity,
                    extract_dir,
                    native_results['elemental_refs'],
                    log=lambda msg: print(f"\033[93m[{impurity}] {msg}\033[0m")
                )
                
                if imp_results:
                    # Calculate Œº(defect) for impurity
                    imp_defect_mu = chempot_calculate_defect_mu_table(
                        imp_results['absolute_mu_df'],
                        native_results['host_elements'],
                        impurity=impurity
                    )
                    imp_results['defect_mu_table'] = imp_defect_mu
                    results_by_impurity[impurity] = imp_results
                    
                    # Display
                    print("\n" + "-"*80)
                    print(f"{impurity}: Absolute Œº (eV)")
                    print(f"Matched formula: {imp_results['matched_formula']}")
                    print("-"*80)
                    display_cols = ['Limit'] + [col for col in imp_results['absolute_mu_df'].columns if col.startswith('mu_')]
                    print(imp_results['absolute_mu_df'][display_cols].to_string(index=False))
                    
                    print("\n" + "-"*80)
                    print(f"{impurity}: Œº(defect) [eV]")
                    print("-"*80)
                    print(imp_defect_mu.to_string(index=False))
                    
            except Exception as e:
                print(f"\033[91m[ERROR] {impurity} calculation failed: {e}\033[0m")
                continue
        
        print("\n" + "="*80)
        print("\033[92m[SUCCESS] Chemical potential calculations complete!\033[0m")
        
        chempot_status.value = ("<div style='padding:10px;background:#f9fafb;border-left:4px solid #9ca3af;"
                               "border-radius:8px;font-size:13px;'><b>Status:</b> Complete! ‚úì</div>")
    
    # Store globally
    chempot_global_results = {
        'composition': composition,
        'impurities': impurities,
        'results_by_impurity': results_by_impurity
    }
    
    # Enable downloads
    chempot_download_native_btn.disabled = False
    chempot_download_impurities_btn.disabled = False if impurities else True
    chempot_download_all_btn.disabled = False

chempot_run_button.on_click(chempot_on_run_clicked)

# Download handlers
def chempot_on_download_native(b):
    if not chempot_global_results:
        return
    
    native_data = chempot_global_results['results_by_impurity']['native']
    canonical = native_data['canonical']
    
    zip_buffer = io.BytesIO()
    with zipfile.ZipFile(zip_buffer, 'w', zipfile.ZIP_DEFLATED) as zf:
        zf.writestr(f"{canonical}_delta_mu.csv", native_data['delta_mu_df'].to_csv(index=False))
        zf.writestr(f"{canonical}_absolute_mu.csv", native_data['absolute_mu_df'].to_csv(index=False))
        zf.writestr(f"{canonical}_defect_mu.csv", native_data['defect_mu_table'].to_csv(index=False))
        
        refs_df = pd.DataFrame([native_data['elemental_refs']]).T
        refs_df.columns = ['Œº¬∞ (eV/atom)']
        zf.writestr(f"{canonical}_elemental_refs.csv", refs_df.to_csv())
        
        readme = f"""
# Native Chemical Potential Results
Composition: {chempot_global_results['composition']}
Canonical: {canonical}
Matched Formula: {native_data['matched_formula']}
Elements: {native_data['host_elements']}

## Files:
- {canonical}_delta_mu.csv: ŒîŒº (eV) for each condition
- {canonical}_absolute_mu.csv: Absolute Œº (eV) = ŒîŒº + Œº¬∞
- {canonical}_defect_mu.csv: Œº(defect) for all native defects
- {canonical}_elemental_refs.csv: Elemental reference energies Œº¬∞ (eV/atom)

## Defect Types:
- V_X: Vacancy on X site ‚Üí Œº = Œº_X
- X_i: X interstitial ‚Üí Œº = -Œº_X
- X_Y: X on Y site (antisite) ‚Üí Œº = Œº_Y - Œº_X
"""
        zf.writestr("README_native.txt", readme)
    
    zip_buffer.seek(0)
    display(chempot_create_download_link(zip_buffer.read(), f"{canonical}_native_chempot.zip", "application/zip"))

def chempot_on_download_impurities(b):
    if not chempot_global_results:
        return
    
    results = chempot_global_results['results_by_impurity']
    canonical = results['native']['canonical']
    
    zip_buffer = io.BytesIO()
    with zipfile.ZipFile(zip_buffer, 'w', zipfile.ZIP_DEFLATED) as zf:
        for imp_name, imp_data in results.items():
            if imp_name == 'native':
                continue
            
            impurity = imp_data['impurity']
            zf.writestr(f"{canonical}_{impurity}_delta_mu.csv", imp_data['delta_mu_df'].to_csv(index=False))
            zf.writestr(f"{canonical}_{impurity}_absolute_mu.csv", imp_data['absolute_mu_df'].to_csv(index=False))
            zf.writestr(f"{canonical}_{impurity}_defect_mu.csv", imp_data['defect_mu_table'].to_csv(index=False))
            
            refs_df = pd.DataFrame([imp_data['elemental_refs']]).T
            refs_df.columns = ['Œº¬∞ (eV/atom)']
            zf.writestr(f"{canonical}_{impurity}_elemental_refs.csv", refs_df.to_csv())
        
        impurities_list = [k for k in results.keys() if k != 'native']
        readme = f"""
# Impurity Chemical Potential Results
Composition: {chempot_global_results['composition']}
Canonical: {canonical}
Impurities: {', '.join(impurities_list)}

## Files per impurity X:
- {canonical}_X_delta_mu.csv: ŒîŒº including impurity
- {canonical}_X_absolute_mu.csv: Absolute Œº for host + impurity
- {canonical}_X_defect_mu.csv: Œº(defect) including X_i and X_Y antisites
- {canonical}_X_elemental_refs.csv: Elemental references including Œº¬∞_X
"""
        zf.writestr("README_impurities.txt", readme)
    
    zip_buffer.seek(0)
    display(chempot_create_download_link(zip_buffer.read(), f"{canonical}_impurities_chempot.zip", "application/zip"))

def chempot_on_download_all(b):
    if not chempot_global_results:
        return
    
    results = chempot_global_results['results_by_impurity']
    canonical = results['native']['canonical']
    
    zip_buffer = io.BytesIO()
    with zipfile.ZipFile(zip_buffer, 'w', zipfile.ZIP_DEFLATED) as zf:
        # Native
        native = results['native']
        zf.writestr(f"native/{canonical}_delta_mu.csv", native['delta_mu_df'].to_csv(index=False))
        zf.writestr(f"native/{canonical}_absolute_mu.csv", native['absolute_mu_df'].to_csv(index=False))
        zf.writestr(f"native/{canonical}_defect_mu.csv", native['defect_mu_table'].to_csv(index=False))
        
        refs_native = pd.DataFrame([native['elemental_refs']]).T
        refs_native.columns = ['Œº¬∞ (eV/atom)']
        zf.writestr(f"native/{canonical}_elemental_refs.csv", refs_native.to_csv())
        
        # Impurities
        for imp_name, imp_data in results.items():
            if imp_name == 'native':
                continue
            
            impurity = imp_data['impurity']
            zf.writestr(f"impurities/{canonical}_{impurity}_delta_mu.csv", imp_data['delta_mu_df'].to_csv(index=False))
            zf.writestr(f"impurities/{canonical}_{impurity}_absolute_mu.csv", imp_data['absolute_mu_df'].to_csv(index=False))
            zf.writestr(f"impurities/{canonical}_{impurity}_defect_mu.csv", imp_data['defect_mu_table'].to_csv(index=False))
            
            refs_imp = pd.DataFrame([imp_data['elemental_refs']]).T
            refs_imp.columns = ['Œº¬∞ (eV/atom)']
            zf.writestr(f"impurities/{canonical}_{impurity}_elemental_refs.csv", refs_imp.to_csv())
        
        # Master README
        impurities_list = [k for k in results.keys() if k != 'native']
        readme = f"""
# Complete Chemical Potential Results
Composition: {chempot_global_results['composition']}
Canonical: {canonical}
Matched Formula: {native['matched_formula']}
Elements: {native['host_elements']}
Impurities: {', '.join(impurities_list) if impurities_list else 'None'}
Date: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}

## Directory Structure:
native/
  - {canonical}_delta_mu.csv
  - {canonical}_absolute_mu.csv
  - {canonical}_defect_mu.csv
  - {canonical}_elemental_refs.csv

impurities/
  - {canonical}_X_delta_mu.csv (for each impurity X)
  - {canonical}_X_absolute_mu.csv
  - {canonical}_X_defect_mu.csv
  - {canonical}_X_elemental_refs.csv

## Defect Chemical Potential Formulas:
V_X (vacancy): Œº = Œº_X
X_i (interstitial): Œº = -Œº_X
X_Y (antisite): Œº = Œº_Y - Œº_X

## Chemical Potential Definition:
ŒîŒº_X = Œº_X - Œº¬∞_X
where Œº¬∞_X is the elemental reference energy per atom

Absolute: Œº_X = ŒîŒº_X + Œº¬∞_X
"""
        zf.writestr("README.txt", readme)
    
    zip_buffer.seek(0)
    display(chempot_create_download_link(zip_buffer.read(), f"{canonical}_ALL_CHEMPOT_RESULTS.zip", "application/zip"))

chempot_download_native_btn.on_click(chempot_on_download_native)
chempot_download_impurities_btn.on_click(chempot_on_download_impurities)
chempot_download_all_btn.on_click(chempot_on_download_all)

# =====================================================================================
# LAYOUT
# =====================================================================================

chempot_instructions = widgets.HTML("""
<div style="background:#f9fafb;padding:15px;border-radius:8px;border-left:4px solid #9ca3af;margin:10px 0;font-size:13px;">
    <b style="color:#4b5563;">üìö How This Tool Works:</b>
    <ol style="margin:5px 0;padding-left:20px;color:#6b7280;line-height:1.8;">
        <li>Select composition from database</li>
        <li>Tool converts to canonical formula (e.g., CdSe0.25Te0.75 ‚Üí Cd25Se3Te22)</li>
        <li><b>Auto-matches to closest available phase</b> (e.g., Cd25Se3Te22 ‚Üí Cd9Te8Se)</li>
        <li>Downloads & filters CompetingPhases (memory-optimized)</li>
        <li>Computes chemical potentials for native + impurities</li>
        <li>Calculates Œº(defect) for all defect types</li>
    </ol>
</div>
""")

chempot_left_panel = widgets.VBox([
    widgets.HTML("<h4 style='color:#4b5563;font-weight:700;margin:5px 0;'>Step 1: Select Composition</h4>"),
    chempot_composition_dropdown,
    widgets.HTML("<h4 style='color:#4b5563;font-weight:700;margin:15px 0 5px 0;'>Step 2: Enter Impurities (Optional)</h4>"),
    chempot_impurities_text,
    widgets.HTML("<h4 style='color:#4b5563;font-weight:700;margin:15px 0 5px 0;'>Step 3: Calculate</h4>"),
    chempot_run_button,
    chempot_progress,
    chempot_status,
], layout=widgets.Layout(width='48%', padding='10px'))

chempot_right_panel = widgets.VBox([
    widgets.HTML("<h4 style='color:#4b5563;font-weight:700;margin:0 0 5px 0;'>üì• Downloads</h4>"),
    chempot_download_native_btn,
    chempot_download_impurities_btn,
    chempot_download_all_btn,
], layout=widgets.Layout(width='48%', padding='10px'))

chempot_main_layout = widgets.VBox([
    chempot_header,
    chempot_instructions,
    widgets.HBox([chempot_left_panel, chempot_right_panel], layout=widgets.Layout(justify_content='space-between')),
    widgets.HTML("<h4 style='color:#4b5563;font-weight:700;margin:15px 0 5px 0;'>üìü Execution Log</h4>"),
    chempot_log_output,
])

# =====================================================================================
# DISPLAY
# =====================================================================================

display(chempot_main_layout)
```
